Section A : Introduction to Constraints

-------------------------------------------------------------------------------------------------------------------------------

1. What are Constraints?

📖 Definition (Interview Ready)

Constraints are rules applied to table columns that ensure the accuracy, consistency, and integrity of data stored in a database. They prevent invalid data from being inserted, updated, or deleted.


-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need Constraints?

Imagine a customer database.

Without Constraints, users could insert:

Duplicate Customer IDs

Negative salaries

Blank employee names

Invalid email addresses

Orders for customers that do not exist

These issues lead to poor data quality.

Constraints help prevent these problems automatically.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Scenario

-------------------------------------------------------------------------------------------------------------------------------

An online shopping company stores customer information.

Every customer must have:

A unique Customer ID

A customer name

A valid email address

Instead of checking these rules in every application, the database enforces them using Constraints.

Section C : PRIMARY KEY

-------------------------------------------------------------------------------------------------------------------------------

2. What is a PRIMARY KEY?

📖 Definition (Interview Ready)

A PRIMARY KEY is a constraint that uniquely identifies each record in a table. It does not allow duplicate values or NULL values.


-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need a PRIMARY KEY?

Without a PRIMARY KEY, we cannot uniquely identify individual records.

Imagine two customers having the same Customer ID.

The database would not know which record belongs to which customer.

Syntax

CREATE TABLE Customers

(

CustomerID INT PRIMARY KEY,

CustomerName VARCHAR(100),

City VARCHAR(50)

);

Business Requirement

Create a Customers table where every customer must have a unique Customer ID.

Query

CREATE TABLE Customers
(

CustomerID INT PRIMARY KEY,

CustomerName VARCHAR(100),

City VARCHAR(50)

);

Insert Valid Data

INSERT INTO Customers

VALUES

(101,'Rahul','Delhi'),

(102,'Priya','Mumbai'),

(103,'Amit','Lucknow');

Output

CustomerID	CustomerName	City

101	Rahul	Delhi

102	Priya	Mumbai

103	Amit	Lucknow

Attempt to Insert Duplicate

INSERT INTO Customers

VALUES
(101,'Neha','Pune');
Result

❌ Error

Violation of PRIMARY KEY constraint.

Cannot insert duplicate key.

Attempt to Insert NULL

INSERT INTO Customers

VALUES

(NULL,'Anita','Jaipur');

Result

❌ Error

PRIMARY KEY columns cannot contain NULL values.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Important Rules

-------------------------------------------------------------------------------------------------------------------------------

A PRIMARY KEY:

✅ Must contain unique values.

✅ Cannot contain NULL values.

✅ Only one PRIMARY KEY is allowed per table.

📖 Definition (Interview Ready)

The NOT NULL constraint ensures that a column cannot contain NULL values. Every row must have a value for that column.


-------------------------------------------------------------------------------------------------------------------------------

(

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a Constraint?

**Answer**

A Constraint is a rule enforced on table columns to maintain the accuracy and integrity of data, such as PRIMARY KEY, FOREIGN KEY, NOT NULL, UNIQUE, CHECK, and DEFAULT.

Q2. What is a PRIMARY KEY?

**Answer**

A PRIMARY KEY is a column (or set of columns) that uniquely identifies each row in a table; it cannot contain NULL values and must be unique.

Q3. Can a PRIMARY KEY contain NULL values?

**Answer**

No. A PRIMARY KEY column cannot contain NULL values.

Q4. Can a table have two PRIMARY KEY constraints?

**Answer**

No, a table can have only one PRIMARY KEY, though that key can be composed of multiple columns (a composite key).

Q5. What is NOT NULL?

**Answer**

NOT NULL is a constraint that ensures a column cannot store a NULL value; every row must have a value for that column.

Q6. What is the difference between PRIMARY KEY and NOT NULL?

**Answer**

A PRIMARY KEY uniquely identifies each row and automatically implies NOT NULL and uniqueness, while a NOT NULL constraint alone only ensures a column has a value without requiring uniqueness.

Section A : UNIQUE Constraint

-------------------------------------------------------------------------------------------------------------------------------

1. What is a UNIQUE Constraint?

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

A UNIQUE constraint ensures that all values in a column are unique. Unlike a PRIMARY KEY, a UNIQUE constraint allows NULL values (SQL Server allows one NULL value).


-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need UNIQUE?

Some columns should never contain duplicate values.

Examples:

Email Address

Aadhaar Number

PAN Number

Passport Number

Username

Although these columns are not Primary Keys, they must still remain unique.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Scenario

An online shopping platform requires every customer to register using a unique email address.

Two customers cannot use the same email.

Syntax

CREATE TABLE Customers
(
CustomerID INT PRIMARY KEY,

CustomerName VARCHAR(100),

Email VARCHAR(100) UNIQUE

);

Example

CREATE TABLE Customers

(

CustomerID INT PRIMARY KEY,

CustomerName VARCHAR(100),

Email VARCHAR(100) UNIQUE


);

INSERT INTO Customers

VALUES
(101,'Rahul','rahul@gmail.com'),

(102,'Priya','priya@gmail.com');

Attempt to Insert Duplicate

INSERT INTO Customers

VALUES

(103,'Amit','rahul@gmail.com');

Result

❌ Error

Violation of UNIQUE constraint.

Cannot insert duplicate value.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Important Note

-------------------------------------------------------------------------------------------------------------------------------

In SQL Server:

A UNIQUE column can contain one NULL value.

Example

INSERT INTO Customers
VALUES
(104,'Neha',NULL);

This is allowed.

A second NULL may violate the UNIQUE constraint depending on the database system and implementation. For SQL Server, only one 
NULL is allowed in a UNIQUE column.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a UNIQUE constraint?

**Answer**

A UNIQUE constraint ensures that all values in a column (or combination of columns) are distinct across the table.

Q2. What is the difference between PRIMARY KEY and UNIQUE?

**Answer**

A table can have only one PRIMARY KEY (which disallows NULLs), while it can have multiple UNIQUE constraints, and a UNIQUE column can generally allow one NULL value depending on the database.

Q3. What is a DEFAULT constraint?

**Answer**

A DEFAULT constraint automatically assigns a specified default value to a column when no value is provided during an INSERT.

Q4. What is a CHECK constraint?

**Answer**

A CHECK constraint restricts the values that can be inserted into a column by enforcing a specified condition, e.g., CHECK (Salary > 0).

Q5. Can a UNIQUE column contain NULL values?

**Answer**

Yes, in most databases a UNIQUE column can contain one NULL value, since NULL is not considered equal to another NULL.

Q6. Why do we use CHECK constraints?

**Answer**

CHECK constraints are used to enforce specific business rules at the database level, such as ensuring salary or age values stay within a valid range.

Q7. What is the difference between CHECK and DEFAULT?

**Answer**

CHECK validates that a value meets a specified condition, while DEFAULT automatically supplies a value when none is provided; DEFAULT does not validate existing data.

1. What is a FOREIGN KEY?

📖 Definition (Interview Ready)

A FOREIGN KEY is a constraint that establishes a relationship between two tables by referencing the PRIMARY KEY of another table. It ensures that data remains consistent and prevents invalid references.

Now the order belongs to no customer.

This creates inconsistent data.

A FOREIGN KEY prevents this problem.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a FOREIGN KEY?

**Answer**

A FOREIGN KEY is a column that references the PRIMARY KEY of another table, creating a relationship between the two tables.

Q2. Why do we use FOREIGN KEY?

**Answer**

FOREIGN KEY is used to enforce referential integrity, ensuring that relationships between tables remain valid and consistent.

Q3. What is Referential Integrity?

**Answer**

Referential Integrity is a database rule ensuring that a foreign key value always matches an existing primary key value in the referenced table (or is NULL).

Q4. What is the difference between PRIMARY KEY and FOREIGN KEY?

**Answer**

A PRIMARY KEY uniquely identifies rows within its own table, while a FOREIGN KEY references the PRIMARY KEY of another table to establish a relationship between the two.

Q5. What is a Parent Table?

**Answer**

A Parent Table is the table that contains the PRIMARY KEY being referenced by a FOREIGN KEY in another table.

Q6. What is a Child Table?

**Answer**

A Child Table is the table that contains the FOREIGN KEY referencing the PRIMARY KEY of the Parent Table.

Q7. What is ON DELETE CASCADE?

**Answer**

ON DELETE CASCADE automatically deletes the corresponding rows in the child table when the referenced row in the parent table is deleted.

Q8. What is ON UPDATE CASCADE?

**Answer**

ON UPDATE CASCADE automatically updates the foreign key value in the child table when the referenced primary key value in the parent table is updated.

Q9. Can a table have multiple FOREIGN KEY constraints?

**Answer**

Yes, a table can have multiple FOREIGN KEY constraints, each referencing a different (or the same) parent table.

Q10. Why are FOREIGN KEY indexes recommended?

**Answer**

Indexing foreign key columns improves the performance of JOIN operations and speeds up enforcement of referential integrity checks during INSERT, UPDATE, and DELETE operations.

Section A : Composite PRIMARY KEY

-------------------------------------------------------------------------------------------------------------------------------

1. What is a Composite PRIMARY KEY?

📖 Definition (Interview Ready)

A Composite PRIMARY KEY is a PRIMARY KEY that consists of two or more columns. The combination of these columns uniquely identifies each row in a table.



-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need a Composite PRIMARY KEY?

Sometimes a single column is not enough to uniquely identify a record.

Example:

A student can enroll in multiple courses.

A course can have multiple students.

Section G : Performance Tips

-------------------------------------------------------------------------------------------------------------------------------

✅ Tip 1

PRIMARY KEY columns are automatically indexed in SQL Server.

✅ Tip 2

Index FOREIGN KEY columns in large tables to improve JOIN performance.

✅ Tip 3

Avoid unnecessary CHECK conditions with complex expressions.

✅ Tip 4

Use DEFAULT to reduce repetitive application logic.

✅ Tip 5

Choose the correct constraint during database design instead of relying on application code.

Section H : Best Practices

-------------------------------------------------------------------------------------------------------------------------------

✅ Every table should have a PRIMARY KEY.

✅ Use FOREIGN KEY to maintain relationships.

✅ Use UNIQUE for alternate identifiers like Email or Username.

✅ Use CHECK to enforce business rules.

✅ Use DEFAULT for common initial values.

✅ Apply NOT NULL to mandatory columns.
