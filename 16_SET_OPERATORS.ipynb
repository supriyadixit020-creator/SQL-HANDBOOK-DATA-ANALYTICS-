Section A : Introduction to Set Operators

-------------------------------------------------------------------------------------------------------------------------------

1. What are Set Operators?

📖 Definition (Interview Ready)

Set Operators are SQL operators used to combine the results of two or more SELECT statements into a single result set. Depending on the operator used, they can merge results, remove duplicates, return only common rows, or identify differences between result sets.

-------------------------------------------------------------------------------------------------------------------------------


Why Do We Need Set Operators?

In real-world databases, related data is often stored in different tables.

Instead of querying each table separately, Set Operators allow us to combine the results into a single report.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a Set Operator?

**Answer**

Set Operators are SQL operators (UNION, UNION ALL, INTERSECT, EXCEPT) used to combine the results of two or more SELECT statements into a single result set.

Q2. What is the difference between UNION and UNION ALL?

**Answer**

UNION combines results and removes duplicate rows, while UNION ALL combines results and keeps all duplicate rows.

Q3. Which Set Operator removes duplicate rows?

**Answer**

UNION, INTERSECT, and EXCEPT all remove duplicate rows automatically; only UNION ALL keeps duplicates.

Q4. Which Set Operator generally performs better?

**Answer**

UNION ALL generally performs better than UNION because it does not need to compare and remove duplicate rows.

Q5. What conditions must be met before using a Set Operator?

**Answer**

Both SELECT statements must return the same number of columns, the corresponding columns must have compatible data types, and the columns must be in the same order.

Section C : INTERSECT vs INNER JOIN

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Both seem to return matching records.

What is the difference?

Feature	INTERSECT	INNER JOIN

Purpose	Returns common rows	Combines related rows

Duplicate Handling	Removes duplicates	Keeps matching rows

Can return columns from both tables	❌	✅

Requires JOIN condition	❌	✅

Example

INTERSECT
SELECT EmployeeID
FROM DelhiEmployees

INTERSECT

SELECT EmployeeID
FROM MumbaiEmployees;

Returns only common Employee IDs.

INNER JOIN

SELECT
d.EmployeeID,
d.EmployeeName,
m.OfficeLocation

FROM DelhiEmployees d

INNER JOIN MumbaiEmployees m
ON d.EmployeeID = m.EmployeeID;

Returns columns from both tables.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

Use INTERSECT when you only need the common rows.

Use INNER JOIN when you need to combine additional information from multiple tables.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is INTERSECT?

**Answer**

INTERSECT returns only the rows that exist in both SELECT statements, removing duplicates automatically.

Q2. What is EXCEPT?

**Answer**

EXCEPT returns the rows from the first SELECT statement that do not exist in the second SELECT statement.

Q3. What is the difference between INTERSECT and INNER JOIN?

**Answer**

INTERSECT compares entire rows from two queries and returns only common rows without combining columns, while INNER JOIN combines columns from two tables based on a matching join condition.

Q4. What is the difference between EXCEPT and NOT EXISTS?

**Answer**

EXCEPT compares two complete result sets and returns rows only in the first, while NOT EXISTS checks row-by-row using a correlated subquery and is generally more flexible for complex filtering conditions.

Q5. Which Set Operator returns only common records?

**Answer**

INTERSECT returns only the records common to both result sets.
