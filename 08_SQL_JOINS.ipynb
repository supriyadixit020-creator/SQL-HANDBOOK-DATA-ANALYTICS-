4. Primary Key and Foreign Key

-------------------------------------------------------------------------------------------------------------------------------

Before learning JOINS, you must understand table relationships.

-------------------------------------------------------------------------------------------------------------------------------

📖 Primary Key

-------------------------------------------------------------------------------------------------------------------------------

Definition

A Primary Key is a column (or set of columns) that uniquely identifies each row in a table.

-------------------------------------------------------------------------------------------------------------------------------

Example

Customers

CustomerID	CustomerName

1	John

2	Emma

3	David

CustomerID is the Primary Key because every value is unique.

-------------------------------------------------------------------------------------------------------------------------------

📖 Foreign Key

-------------------------------------------------------------------------------------------------------------------------------


Definition

A Foreign Key is a column that references the Primary Key of another table, creating a relationship between the tables.

-------------------------------------------------------------------------------------------------------------------------------

Orders

OrderID	CustomerID	Amount

101	1	500

102	2	800

103	1	600

Here:

CustomerID

is the Foreign Key because it references:

Customers.CustomerID

6. Types of SQL JOINS

-------------------------------------------------------------------------------------------------------------------------------


SQL provides several types of joins.

-------------------------------------------------------------------------------------------------------------------------------

Join Type	Purpose

INNER JOIN	Returns matching rows from both tables

LEFT JOIN	Returns all rows from the left table and matching rows from the right table

RIGHT JOIN	Returns all rows from the right table and matching rows from the left table

FULL OUTER JOIN	Returns all rows from both tables, matching where possible

CROSS JOIN	Returns every possible combination of rows

SELF JOIN	Joins a table with itself

-------------------------------------------------------------------------------------------------------------------------------

📌 Interview Tip: INNER JOIN and LEFT JOIN are the most commonly used joins in Data Analytics.

Q1. What is a SQL JOIN?

Answer

A SQL JOIN combines rows from two or more related tables based on a common column, usually a Primary Key and a Foreign Key.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Why are SQL JOINS used?

Answer

SQL JOINS are used to retrieve related data stored across multiple tables and present it as a single result set.

-------------------------------------------------------------------------------------------------------------------------------

Q3. What is the difference between a Primary Key and a Foreign Key?

A Primary Key uniquely identifies each row in a table, while a Foreign Key references the Primary Key of another table to establish a relationship.

Answer

INNER JOIN and LEFT JOIN are the most commonly used joins in reporting and analytics.

Definition (Interview Ready)

A LEFT JOIN returns all rows from the left table and the matching rows from the right table. If no matching row exists in the right table, SQL returns NULL for the right table's columns.

-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need LEFT JOIN?

Suppose we have these tables.

Customers

CustomerID	CustomerName

1	John

2	Emma

3	David

4	Sophia

Orders

OrderID	CustomerID	Amount

101	1	500

102	2	1200

103	1	900

-------------------------------------------------------------------------------------------------------------------------------

Business Question:

-------------------------------------------------------------------------------------------------------------------------------

Show all customers, even those who have never placed an order.

INNER JOIN cannot answer this because it only returns matching rows.

LEFT JOIN is the correct choice.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

Businesses often want to identify:

Customers who never placed an order.

Employees without projects.

Students who haven't enrolled in any course.

Products that have never been sold.

These reports require a LEFT JOIN.

FROM LeftTable

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a LEFT JOIN?
Answer

A LEFT JOIN returns all rows from the left table and the matching rows from the right table. If no match exists, SQL returns NULL values for the right table's columns.

-------------------------------------------------------------------------------------------------------------------------------

Q2. What is the difference between INNER JOIN and LEFT JOIN?
Answer

INNER JOIN returns only matching rows, whereas LEFT JOIN returns all rows from the left table and matching rows from the right table.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Why do NULL values appear in a LEFT JOIN?
Answer

NULL values appear when a row from the left table has no matching row in the right table.

-------------------------------------------------------------------------------------------------------------------------------

Q4. How can you find records without a match?
Answer

Perform a LEFT JOIN and filter using:

WHERE RightTable.Column IS NULL;
