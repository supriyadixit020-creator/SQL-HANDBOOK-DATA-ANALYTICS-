1. What is NULl?

📖 Definition (Interview Ready)

NULL represents a missing, unknown, or unavailable value in a database. It does not mean zero, an empty string, or a blank space.


------------------------------------------------------------------------------------------------------------------------------

2. NULL vs 0 vs Empty String

------------------------------------------------------------------------------------------------------------------------------

Value	Meaning
NULL	Unknown or Missing Value
0	Actual Numeric Zero
''	Empty Text (No Characters)
' '	A Space Character
Example

Employee	Bonus

John	5000

Emma	NULL

David	0

Meaning:

John received ₹5,000.

Emma's bonus is unknown or not recorded.

David received ₹0 as bonus.

------------------------------------------------------------------------------------------------------------------------------

📌 NULL and 0 are completely different values.

------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

Common NULL values:

Phone Number

Email Address

Delivery Date

Bonus

Discount

Manager ID

Resignation Date

3. Why NULL is Important

Consider this query:

SELECT
EmployeeName,
Salary + Bonus AS Total_Salary
FROM Employees;

Output

Employee	Total Salary

John	55000

Emma	NULL

David	48000

Sophia	NULL

Explanation

Any arithmetic operation involving NULL returns NULL.

Examples:

100 + NULL = NULL

500 × NULL = NULL

1000 - NULL = NULL

------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

Any mathematical calculation with NULL returns NULL.

This is one of the most frequently asked SQL interview concepts.

Section A : ISNULL()

------------------------------------------------------------------------------------------------------------------------------

4. ISNULL()

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The ISNULL() function replaces a NULL value with a specified replacement value in SQL Server.

------------------------------------------------------------------------------------------------------------------------------

Syntax

ISNULL(expression, replacement_value)

------------------------------------------------------------------------------------------------------------------------------

Parameters

Parameter	Meaning

expression	Value to check

replacement_value	Value returned if expression is NULL

Business Question


Replace missing bonuses with 0.

Query

SELECT
EmployeeName,
Bonus,
ISNULL(Bonus,0) AS Bonus_After_Replacement
FROM Employees;

Output

Employee	Bonus	Bonus_After_Replacement

John	5000	5000

Emma	NULL	0

David	3000	3000

Sophia	NULL	0

Explanation

If Bonus is NULL

↓

Return 0

Otherwise

↓

Return Bonus

💼 Business Use Cases

Missing bonuses

Missing discounts

Missing salaries

Missing stock quantities

Missing customer ratings

------------------------------------------------------------------------------------------------------------------------------

5. ISNULL() in Calculations

Business Question

Calculate Total Salary.

Query

SELECT
EmployeeName,
Salary,
Bonus,
Salary + ISNULL(Bonus,0) AS Total_Salary
FROM Employees;

Output

Employee	Salary	Bonus	Total Salary

John	50000	5000	55000

Emma	60000	NULL	60000

David	45000	3000	48000

Sophia	55000	NULL	55000

Explanation

Instead of adding NULL,

SQL adds:

Salary + 0

which produces the expected result.

7. Best Practices


✅ Replace NULL only when it makes business sense.

✅ Use meaningful replacement values.

Examples:

Bonus → 0

Phone → 'Not Available'

City → 'Unknown'

✅ Document why NULL values are replaced.

------------------------------------------------------------------------------------------------------------------------------

------------------------------------------------------------------------------------------------------------------------------

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

------------------------------------------------------------------------------------------------------------------------------

------------------------------------------------------------------------------------------------------------------------------

Section A : COALESCE()

------------------------------------------------------------------------------------------------------------------------------

1. COALESCE()


📖 Definition (Interview Ready)

COALESCE() returns the first non-NULL value from a list of expressions.

------------------------------------------------------------------------------------------------------------------------------

Syntax

COALESCE(value1, value2, value3, ...)

------------------------------------------------------------------------------------------------------------------------------

Business Question

Display the employee's preferred contact.

Priority:

Phone

Email

"Not Available"

Query

SELECT
    EmployeeName,
    COALESCE(Phone, Email, 'Not Available') AS Preferred_Contact
FROM Employees;

Output

Employee	Preferred_Contact

John	9876543210

Emma	emma@gmail.com

David	9876543211

Sophia	Not Available

Explanation

SQL checks values from left to right.

For Emma:

Phone = NULL

↓

Email = emma@gmail.com

↓

Return Email

For Sophia:

Phone = NULL

↓

Email = NULL

↓

Return "Not Available"

💼 Business Use Cases

Customer Contact Information

Shipping Address

Emergency Contact

Backup Phone Number

Multiple Email Addresses

------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

COALESCE() can check multiple values, whereas ISNULL() checks only one expression.

------------------------------------------------------------------------------------------------------------------------------

2. ISNULL() vs COALESCE()

------------------------------------------------------------------------------------------------------------------------------

Feature	ISNULL()	COALESCE()

SQL Server	✅	✅

SQL Standard	❌	✅

Multiple Values	❌	✅

Returns First Non-NULL	❌	✅

Example

ISNULL()
SELECT
ISNULL(Bonus,0)
FROM Employees;
COALESCE()
SELECT
COALESCE(Bonus,Commission,Incentive,0)
FROM Employees;
Business Insight

If you have multiple possible values,

always prefer

COALESCE()

Section B : NULLIF()

------------------------------------------------------------------------------------------------------------------------------

3. NULLIF()

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

NULLIF() compares two expressions and returns NULL if they are equal; otherwise, it returns the first expression.


------------------------------------------------------------------------------------------------------------------------------

Syntax

NULLIF(expression1, expression2)

------------------------------------------------------------------------------------------------------------------------------

Example

SELECT
NULLIF(100,100) AS Result;

Output

NULL

Another Example

SELECT
NULLIF(100,200) AS Result;

Output

100

💼 Business Use Cases

Prevent divide-by-zero errors.

Convert placeholder values to NULL.

Data cleaning.

Example


Suppose

Discount
0

5

10

Business Rule:

Treat 0 as missing.

Query

SELECT
Discount,
NULLIF(Discount,0) AS Clean_Discount
FROM Products;

Output

Discount	Clean_Discount

0	NULL

5	5

10	10

IS NOT NULL is used to retrieve rows where a column contains a valid (non-NULL) value.

⚡ Performance Tip

COALESCE() is highly flexible and portable across databases, but if you're writing SQL Server-only code and replacing a single NULL value, ISNULL() is often simpler and easier to read.

------------------------------------------------------------------------------------------------------------------------------

🎤 Interview Questions

------------------------------------------------------------------------------------------------------------------------------

Q1. What does COALESCE() do?

Answer

COALESCE() returns the first non-NULL value from a list of expressions.

------------------------------------------------------------------------------------------------------------------------------

Q2. What is the difference between ISNULL() and COALESCE()?

Answer

ISNULL() replaces a NULL value in one expression, whereas COALESCE() evaluates multiple expressions and returns the first non-NULL value.

------------------------------------------------------------------------------------------------------------------------------

Q3. What does NULLIF() do?

Answer

NULLIF() returns NULL when two expressions are equal; otherwise, it returns the first expression.

------------------------------------------------------------------------------------------------------------------------------

Q4. Why can't we write WHERE Bonus = NULL?

Answer

Because NULL represents an unknown value. SQL requires IS NULL or IS NOT NULL to test for NULL values.

------------------------------------------------------------------------------------------------------------------------------

✅ NULLIF() returns NULL when two values are equal.

✅ IS NULL finds missing values.

✅ IS NOT NULL finds available values.

✅ Never compare NULL using =.

Section A : NULL with Aggregate Functions

------------------------------------------------------------------------------------------------------------------------------

1. NULL with SUM()

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The SUM() function ignores NULL values and adds only non-NULL values.


------------------------------------------------------------------------------------------------------------------------------

Business Question

------------------------------------------------------------------------------------------------------------------------------

Find the total bonus paid.

Query

SELECT
SUM(Bonus) AS Total_Bonus
FROM Employees;

Output

Total_Bonus

8000

Explanation

SQL ignores NULL values.

Calculation:

5000

+

3000

=

8000

💼 Business Insight

This is useful when:

Some employees have not received bonuses.

Missing values should not affect the total.

-------------------------------------------------------------------------------------------------------------------------------

2. NULL with AVG()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The AVG() function ignores NULL values and calculates the average using only non-NULL values.

-------------------------------------------------------------------------------------------------------------------------------

Query

SELECT
AVG(Bonus) AS Average_Bonus
FROM Employees;
Output
Average_Bonus
4000
Explanation

SQL calculates

(5000 + 3000)

÷

2

NOT

÷ 4

because NULL values are ignored.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

AVG() ignores NULL values.

It only considers available values.

-------------------------------------------------------------------------------------------------------------------------------

3. NULL with COUNT()

-------------------------------------------------------------------------------------------------------------------------------

There are three important ways to use COUNT().

COUNT(*)

Returns all rows.

SELECT COUNT(*) AS Total_Rows
FROM Employees;

Output

4
COUNT(Bonus)

Counts only non-NULL values.

SELECT COUNT(Bonus) AS Employees_With_Bonus
FROM Employees;

Output

2
COUNT(EmployeeID)

Counts all non-NULL Employee IDs.

SELECT COUNT(EmployeeID)

FROM Employees;

Output

4


⭐ COUNT Comparison

Function	NULL Counted?

COUNT(*)	✅ Yes

COUNT(Column)	❌ No

SUM()	❌ No

AVG()	❌ No

💼 Business Insight

One of the most common interview questions is:

"What is the difference between COUNT(*) and COUNT(column_name)?"

Interview Answer

COUNT(*) counts every row in the table, whereas COUNT(column_name) counts only rows where that column contains a non-NULL value.

-------------------------------------------------------------------------------------------------------------------------------

Emma	IT

David	NULL

Sophia	HR

Explanation

LEFT JOIN keeps all employees.

If no matching department exists,

DepartmentName becomes NULL.

9. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Thinking AVG() divides by all rows.

Wrong.

It divides only by non-NULL values.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Expecting INNER JOIN to include NULL keys.

INNER JOIN only returns matching rows.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Checking NULL with

= NULL

Always use

IS NULL

-------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Tip

Instead of repeatedly handling NULL values in every report, consider cleaning or standardizing missing values during the ETL process when appropriate. This keeps reporting queries simpler and easier to maintain.


-------------------------------------------------------------------------------------------------------------------------------

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. Does SUM() include NULL values?

Answer

No. SUM() ignores NULL values.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Does AVG() include NULL values?

Answer

No. AVG() ignores NULL values and calculates the average using only non-NULL values.

-------------------------------------------------------------------------------------------------------------------------------

Q3. What is the difference between COUNT(*) and COUNT(column_name)?

Answer

COUNT(*) counts all rows, whereas COUNT(column_name) counts only non-NULL values in that column.

-------------------------------------------------------------------------------------------------------------------------------

Q4. What happens to NULL values in an INNER JOIN?

Answer

Rows with NULL join keys do not match and are therefore excluded from the INNER JOIN result.
