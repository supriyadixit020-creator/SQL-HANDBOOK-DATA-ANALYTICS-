Section A : Introduction

------------------------------------------------------------------------------------------------------------------------------

1. What are Window Functions?

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

------------------------------------------------------------------------------------------------------------------------------

Window Functions perform calculations across a set of related rows while returning each individual row. Unlike GROUP BY, they do not collapse multiple rows into a single row.



------------------------------------------------------------------------------------------------------------------------------

🎯 Easy Visualization

------------------------------------------------------------------------------------------------------------------------------

Employees

↓

Window Function

↓

Calculates Result

↓

Keeps Every Row

↓

Displays Final Output

------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Need Window Functions?

------------------------------------------------------------------------------------------------------------------------------

Suppose you want to calculate the average salary.

Using GROUP BY

SELECT
Department,
AVG(Salary)
FROM Employees
GROUP BY Department;

Output

Department	Average Salary

HR	52500

IT	75000

Finance	45000

The individual employee rows disappear.

Suppose you want:

Employee	Department	Salary	Department Average

John	HR	50000	52500

Michael	HR	55000	52500

Emma	IT	70000	75000

Sophia	IT	80000	75000

This is not possible with GROUP BY alone.

Window Functions solve this problem.

💼 Business Insight

Window Functions are ideal when you want to:

Keep every row.

Compare rows.

Rank rows.

Calculate running totals.

Calculate moving averages.

Analyze trends.

Section B : OVER()

------------------------------------------------------------------------------------------------------------------------------

3. What is OVER()?

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The OVER() clause defines the window (set of rows) on which a Window Function performs its calculation.


------------------------------------------------------------------------------------------------------------------------------

Syntax

------------------------------------------------------------------------------------------------------------------------------

Function_Name() OVER ()

------------------------------------------------------------------------------------------------------------------------------

Example
Business Question

Display every employee along with the company's average salary.

Query

SELECT
EmployeeName,
Salary,
AVG(Salary) OVER() AS CompanyAverage
FROM Employees;

Output

Employee	Salary	CompanyAverage

John	50000	60000

Emma	70000	60000

David	45000	60000

Sophia	80000	60000

Michael	55000	60000

Explanation

AVG(Salary) calculates:

60000

OVER() displays that value on every row instead of returning only one row.

💼 Business Use Cases

Company Average Salary

Overall Sales

Total Revenue

Dashboard KPIs

Section C : PARTITION BY

------------------------------------------------------------------------------------------------------------------------------

4. What is PARTITION BY?

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

PARTITION BY divides the result set into groups, and the Window Function performs calculations separately within each group.


------------------------------------------------------------------------------------------------------------------------------

Syntax

------------------------------------------------------------------------------------------------------------------------------

Function_Name() OVER
(
PARTITION BY Column_Name
)

------------------------------------------------------------------------------------------------------------------------------

Business Question

Display each employee with the average salary of their department.

Query

SELECT
EmployeeName,
Department,
Salary,
AVG(Salary) OVER
(
PARTITION BY Department
) AS DepartmentAverage
FROM Employees;

Output

Employee	Department	Salary	DepartmentAverage

John	HR	50000	52500

Michael	HR	55000	52500

Emma	IT	70000	75000

Sophia	IT	80000	75000

David	Finance	45000	45000

Explanation

Window is divided into:

HR

↓

Average = 52500

--------------------

IT

↓

Average = 75000

--------------------

Finance

↓

Average = 45000

Each employee receives the average of their own department.

Section D : ORDER BY inside OVER()

------------------------------------------------------------------------------------------------------------------------------

5. ORDER BY in Window Functions

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

ORDER BY inside the OVER() clause defines the order in which the Window Function processes rows within each window.

------------------------------------------------------------------------------------------------------------------------------

Syntax

------------------------------------------------------------------------------------------------------------------------------

Function_Name() OVER
(
ORDER BY Column_Name
)

------------------------------------------------------------------------------------------------------------------------------

Example

SELECT
EmployeeName,
Salary,
ROW_NUMBER() OVER
(
ORDER BY Salary DESC
) AS RowNum
FROM Employees;

(We'll study ROW_NUMBER() in detail in Part 2. For now, focus on how ORDER BY controls the order of calculation.)

------------------------------------------------------------------------------------------------------------------------------

6. OVER() vs GROUP BY

------------------------------------------------------------------------------------------------------------------------------

Feature	OVER()	GROUP BY

Keeps Individual Rows	✅	❌

Returns Aggregate Values	✅	✅

Collapses Rows	❌	✅

Used with Window Functions	✅	❌


💼 Business Insight

If you need both detailed rows and summary values together, use Window Functions instead of GROUP BY.

------------------------------------------------------------------------------------------------------------------------------

9. Common Mistakes

------------------------------------------------------------------------------------------------------------------------------


❌ Mistake 1

------------------------------------------------------------------------------------------------------------------------------

Thinking PARTITION BY behaves exactly like GROUP BY.

PARTITION BY keeps all rows, while GROUP BY combines rows.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Forgetting that OVER() is required for Window Functions.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Confusing ORDER BY inside OVER() with the final ORDER BY of the query.

ORDER BY inside OVER() controls the Window Function.
Final ORDER BY controls the displayed output.

-------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Tip

Window Functions are often more efficient and easier to read than complex self-joins used for ranking or running calculations. However, they may require sorting, so indexing relevant columns (such as those used in PARTITION BY or ORDER BY) can improve performance.

-------------------------------------------------------------------------------------------------------------------------------

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a Window Function?

Answer

A Window Function performs calculations across a set of related rows while preserving each individual row in the result set.

-------------------------------------------------------------------------------------------------------------------------------

Q2. What does OVER() do?

Answer

The OVER() clause defines the set of rows (window) on which the Window Function performs its calculation.

-------------------------------------------------------------------------------------------------------------------------------

Q3. What is PARTITION BY?

Answer

PARTITION BY divides the result set into groups so the Window Function calculates separately for each group.

-------------------------------------------------------------------------------------------------------------------------------

Q4. What is the difference between GROUP BY and PARTITION BY?

Answer

GROUP BY combines rows into a single result per group, whereas PARTITION BY keeps all rows and performs calculations within each group.

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

✅ OVER() defines the calculation window.

✅ PARTITION BY divides rows into groups.

✅ ORDER BY inside OVER() controls calculation order.

✅ GROUP BY summarizes rows, while Window Functions preserve them.

Section A : ROW_NUMBER()

-------------------------------------------------------------------------------------------------------------------------------

1. What is ROW_NUMBER()?

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

ROW_NUMBER() is a Window Function that assigns a unique sequential number to each row based on the order specified in the OVER() clause.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Easy Visualization

-------------------------------------------------------------------------------------------------------------------------------

Salary (DESC)

80000

↓

Row Number = 1

------------------

70000

↓

Row Number = 2

------------------

55000

↓

Row Number = 3

------------------

50000

↓

Row Number = 4

------------------

45000

↓

Row Number = 5

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

ROW_NUMBER() OVER
(
ORDER BY Column_Name
)

-------------------------------------------------------------------------------------------------------------------------------

Important Rule

-------------------------------------------------------------------------------------------------------------------------------

ROW_NUMBER() always requires an ORDER BY clause inside OVER().

Without ORDER BY, SQL Server returns an error.

-------------------------------------------------------------------------------------------------------------------------------

Section B : Basic Example

-------------------------------------------------------------------------------------------------------------------------------

Assign a rank to employees based on salary.

Query

Salary,
ROW_NUMBER() OVER
(
ORDER BY Salary DESC
) AS RowNumber
FROM Employees;

Employee	Salary	RowNumber

Sophia	80000	1

Emma	70000	2

Michael	55000	3

John	50000	4

David	45000	5

Explanation

SQL first sorts employees by salary (highest to lowest).

Then assigns numbers:

1

↓

2

↓

3

↓

4

↓

5

Every row gets a unique number.

-------------------------------------------------------------------------------------------------------------------------------

Keep only one record for each email.

Query

WITH DuplicateData AS
(
SELECT *,
ROW_NUMBER() OVER
(

ORDER BY CustomerID
) AS RowNum
FROM Customers
)

SELECT *
FROM DuplicateData
WHERE RowNum = 1;

Output

CustomerID	CustomerName	Email

1	Rahul	rahul@gmail.com

3	Anita	anita@gmail.com

Explanation

The first record gets RowNum = 1.

Duplicate records receive 2, 3, ...

Filtering RowNum = 1 keeps only one record.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is ROW_NUMBER()?

Answer

ROW_NUMBER() is a Window Function that assigns a unique sequential number to each row based on the order specified in the OVER() clause.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Is ORDER BY mandatory with ROW_NUMBER()?

Answer

Yes. ROW_NUMBER() requires an ORDER BY clause inside the OVER() clause.

-------------------------------------------------------------------------------------------------------------------------------

Q3. What does PARTITION BY do with ROW_NUMBER()?

Answer

PARTITION BY restarts the row numbering for each group.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Can ROW_NUMBER() assign the same number to two rows?

Answer

No. ROW_NUMBER() always assigns a unique number to each row, even when values are tied.

✅ It does not assign the same rank to tied values.

RANK() is a Window Function that assigns the same rank to rows with equal values. If there is a tie, the next rank is skipped.

John	70000	4

2. What is LAST_VALUE()?

📖 Definition (Interview Ready)

LAST_VALUE() returns the last value in the current window.


-------------------------------------------------------------------------------------------------------------------------------

⚠️ Most Important Interview Trap

-------------------------------------------------------------------------------------------------------------------------------

Many beginners think:

LAST_VALUE()

returns the last value of the entire table.

❌ Wrong

By default it returns the last value of the current window, which often means the current row.

Wrong Query

SELECT
Month,
Sales,

LAST_VALUE(Sales)
OVER
(
ORDER BY Month
) AS LastSales

FROM MonthlySales;

Wrong Output

Month	LastSales

January	10000

February	12000

March	15000

April	14000

May	18000

It simply repeats the current row because of the default window frame.

✅ Correct Query


SELECT
Month,
Sales,

LAST_VALUE(Sales)
OVER
(
ORDER BY Month
ROWS BETWEEN UNBOUNDED PRECEDING
AND UNBOUNDED FOLLOWING
) AS LastMonthSales

FROM MonthlySales;

Correct Output

Month	LastMonthSales

January	18000

February	18000

March	18000

April	18000

May	18000

If you use LAST_VALUE(), always think about the Window Frame.

This is one of the most common SQL interview traps.

5. SQL Best Practices

-------------------------------------------------------------------------------------------------------------------------------

✅ Write SQL keywords in uppercase.

SELECT
FROM
WHERE
GROUP BY
ORDER BY

✅ Use meaningful aliases.

EmployeeSalary
MonthlySales
DepartmentSummary

✅ Format SQL consistently.

✅ Keep each query focused on one business problem.

✅ Use comments to explain complex logic.

-- Calculate running total

✅ Prefer CTEs over deeply nested subqueries when readability improves.

6. SQL Coding Standards

-------------------------------------------------------------------------------------------------------------------------------

Good
SELECT
    EmployeeName,
    Salary
FROM Employees
WHERE Salary > 50000;
Avoid
select employeename,salary from employees where salary>50000

7. Common SQL Interview Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Forgetting ORDER BY with ranking functions.

❌ Using WHERE instead of HAVING after GROUP BY.

❌ Forgetting PARTITION BY when ranking within groups.

❌ Using = with a Multi-Row Subquery.

❌ Not handling NULL values.

❌ Forgetting aliases for Derived Tables.

❌ Writing unreadable nested queries.

9. SQL Performance Tips

-------------------------------------------------------------------------------------------------------------------------------

✅ Filter data as early as possible.

✅ Select only required columns.

Avoid:

SELECT *

Use:

SELECT EmployeeName, Salary

✅ Create indexes on frequently filtered or sorted columns.

✅ Avoid unnecessary subqueries.

✅ Prefer Window Functions over self-joins for ranking and running calculations.

✅ Review execution plans for large datasets.
