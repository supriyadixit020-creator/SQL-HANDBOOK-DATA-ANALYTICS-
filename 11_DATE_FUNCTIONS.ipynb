Section A : GETDATE()

------------------------------------------------------------------------------------------------------------------------------

3. GETDATE()

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

GETDATE() returns the current system date and time in SQL Server.

------------------------------------------------------------------------------------------------------------------------------

Syntax

GETDATE()

------------------------------------------------------------------------------------------------------------------------------

Business Question

Display today's date and time.

Query

SELECT GETDATE() AS Current_DateTime;

Sample Output

Current_DateTime

2026-07-10 10:35:42.357

Explanation


The function returns:

Current Date ✅

Current Time ✅

based on the SQL Server system clock.

------------------------------------------------------------------------------------------------------------------------------

💼 Business Use Cases

Report generation

Audit logs

Invoice creation

Order timestamps

Login tracking

------------------------------------------------------------------------------------------------------------------------------

4. CAST(GETDATE() AS DATE)

------------------------------------------------------------------------------------------------------------------------------

📖 Definition

CAST(GETDATE() AS DATE) returns only the date portion by removing the time.

------------------------------------------------------------------------------------------------------------------------------

Query

SELECT CAST(GETDATE() AS DATE) AS Today_Date;

Sample Output

Today_Date

2026-07-10

Explanation

GETDATE()

returns

2026-07-10 10:35:42

After casting:

2026-07-10


------------------------------------------------------------------------------------------------------------------------------

5. CURRENT_TIMESTAMP

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

CURRENT_TIMESTAMP returns the current system date and time according to the SQL standard.

------------------------------------------------------------------------------------------------------------------------------

Query

SELECT CURRENT_TIMESTAMP AS Current_DateTime;

Sample Output

Current_DateTime

2026-07-10 10:35:42.357

Database Compatibility

Database	Current Date & Time Function

SQL Server	GETDATE()

SQL Server (Standard SQL)	CURRENT_TIMESTAMP

MySQL	NOW()

PostgreSQL	CURRENT_TIMESTAMP

------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

------------------------------------------------------------------------------------------------------------------------------

If the interviewer asks:

Which function returns the current date and time in SQL Server?

Answer:

GETDATE()

You can also mention:

CURRENT_TIMESTAMP

because it is SQL-standard and supported by SQL Server.

------------------------------------------------------------------------------------------------------------------------------

6. GETDATE() vs CURRENT_TIMESTAMP

------------------------------------------------------------------------------------------------------------------------------

Feature	GETDATE()	CURRENT_TIMESTAMP

Returns Time	✅	✅

9. Best Practices

------------------------------------------------------------------------------------------------------------------------------

✅ Use GETDATE() for SQL Server projects.

✅ Use CURRENT_TIMESTAMP when writing portable SQL.

✅ Store dates using proper DATE or DATETIME data types rather than text.

------------------------------------------------------------------------------------------------------------------------------

10. Common Mistakes

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

------------------------------------------------------------------------------------------------------------------------------

Storing dates as text (VARCHAR).

Always use appropriate date data types.

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

------------------------------------------------------------------------------------------------------------------------------

Using GETDATE() when only the date is needed.

Use:

CAST(GETDATE() AS DATE)

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

------------------------------------------------------------------------------------------------------------------------------

Confusing GETDATE() with NOW().

Remember:

SQL Server → GETDATE()

MySQL → NOW()

------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Note

GETDATE() is lightweight and efficient. However, avoid applying functions to indexed date columns in filtering conditions when a direct comparison can be used, as this may reduce index efficiency.

The DATEPART() function returns a specified part of a date, such as year, month, day, quarter, week, hour, or minute.

8. Best Practices

------------------------------------------------------------------------------------------------------------------------------

✅ Use YEAR(), MONTH(), and DAY() when only one part of the date is required.

✅ Give meaningful aliases.

9. Common Mistakes

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

------------------------------------------------------------------------------------------------------------------------------

Using MONTH() when the business requires the month name.

Use FORMAT() (covered later) if you need names like "January".

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

------------------------------------------------------------------------------------------------------------------------------

Confusing DAY() with the day of the week.

DAY() returns the day of the month, not Monday, Tuesday, etc.

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

------------------------------------------------------------------------------------------------------------------------------

Using DATEPART() with an incorrect date part keyword.

Always verify the supported keywords for your database.

------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Note

Functions like YEAR() or MONTH() in a WHERE clause may prevent efficient use of indexes. When filtering large datasets, direct date range comparisons are often more efficient.

Section A : DATEADD()

------------------------------------------------------------------------------------------------------------------------------

1. DATEADD()

------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The DATEADD() function adds or subtracts a specified number of date parts (days, months, years, etc.) to a given date.

------------------------------------------------------------------------------------------------------------------------------

Syntax

DATEADD(date_part, number, date)

------------------------------------------------------------------------------------------------------------------------------

Parameters

Parameter	Meaning

date_part	Day, Month, Year, etc.

number	Positive = Add, Negative = Subtract

date	Original date

------------------------------------------------------------------------------------------------------------------------------

Business Question

Add 30 days to every employee's joining date.

Query

SELECT
    EmployeeName,
    HireDate,
    DATEADD(DAY,30,HireDate) AS Review_Date
FROM Employees;

Output

EmployeeName	HireDate	Review_Date

John	2021-06-15	2021-07-15

Emma	2020-03-10	2020-04-09

David	2023-01-20	2023-02-19

Sophia	2022-11-05	2022-12-05


Explanation

SQL adds 30 days to each hire date.

------------------------------------------------------------------------------------------------------------------------------

💼 Business Use Cases

Employee probation period

Delivery due dates

Subscription expiry

Invoice due dates

------------------------------------------------------------------------------------------------------------------------------

Example 2

Add one year.

SELECT
EmployeeName,
DATEADD(YEAR,1,HireDate) AS Next_Review
FROM Employees;

------------------------------------------------------------------------------------------------------------------------------

Example 3

Subtract 15 days.

SELECT
DATEADD(DAY,-15,GETDATE()) AS Previous_Date;

Hire Date

↓

💼 Business Use Cases

------------------------------------------------------------------------------------------------------------------------------

Example 2

Calculate order age.

SELECT
OrderID,

Section B : Date Filtering

------------------------------------------------------------------------------------------------------------------------------

4. Filter Current Year's Orders

------------------------------------------------------------------------------------------------------------------------------

Business Question

Display only orders placed in the current year.

Query

SELECT
    *
FROM Orders
WHERE YEAR(OrderDate)=YEAR(GETDATE());

Explanation

The query compares:

Current Year

↓

Order Year

and returns only matching rows.

------------------------------------------------------------------------------------------------------------------------------

⚠️ Better Performance Method

------------------------------------------------------------------------------------------------------------------------------

Instead of:

WHERE YEAR(OrderDate)=2026;

Prefer:

WHERE OrderDate >= '2026-01-01'
AND OrderDate < '2027-01-01';

Why?

The second query is generally more efficient because it can make better use of indexes on OrderDate.

------------------------------------------------------------------------------------------------------------------------------

5. Filter Current Month

------------------------------------------------------------------------------------------------------------------------------

Query

SELECT *
FROM Orders
WHERE MONTH(OrderDate)=MONTH(GETDATE())
AND YEAR(OrderDate)=YEAR(GETDATE());

6. Orders in Last 30 Days

------------------------------------------------------------------------------------------------------------------------------

Query

Explanation

3. Common Interview Traps

-------------------------------------------------------------------------------------------------------------------------------

❌ Trap 1

-------------------------------------------------------------------------------------------------------------------------------

Confusing GETDATE() with CURRENT_TIMESTAMP.

Remember:

GETDATE() → SQL Server specific

CURRENT_TIMESTAMP → SQL Standard

-------------------------------------------------------------------------------------------------------------------------------


❌ Trap 2

-------------------------------------------------------------------------------------------------------------------------------


Using YEAR() in large-table filtering.

Wrong

WHERE YEAR(OrderDate)=2026;

Better

WHERE OrderDate >= '2026-01-01'

AND OrderDate < '2027-01-01';

-------------------------------------------------------------------------------------------------------------------------------

❌ Trap 3

-------------------------------------------------------------------------------------------------------------------------------

Using DATEDIFF(YEAR,...) as an exact age calculation.

It counts year boundaries crossed, which may not equal the exact completed age.

-------------------------------------------------------------------------------------------------------------------------------


❌ Trap 4

-------------------------------------------------------------------------------------------------------------------------------


Thinking FORMAT() changes stored values.

It only changes the display.

-------------------------------------------------------------------------------------------------------------------------------


❌ Trap 5

-------------------------------------------------------------------------------------------------------------------------------

Manually calculating month-end dates.

Always use:

EOMONTH()
