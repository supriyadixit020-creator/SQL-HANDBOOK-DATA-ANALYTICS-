1. Definition (Interview Ready)

The HAVING clause is used to filter groups after the GROUP BY clause has created them. It is commonly used with Aggregate Functions.

-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Need HAVING?

Suppose we have already grouped employees by department.

Now the HR Manager asks:

Show only departments with more than 1 employee.

Show departments whose average salary is greater than ₹50,000.

Show departments whose total salary exceeds ₹1,00,000.

Can we use WHERE?

❌ No.

Because WHERE filters rows before grouping.

We need HAVING to filter the groups after aggregation.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

Imagine a company with 50 departments.

The HR Director wants to see only those departments where the average salary is above ₹60,000.

Instead of manually checking each department, a single HAVING clause generates the report.

-------------------------------------------------------------------------------------------------------------------------------

3. WHERE vs HAVING

-------------------------------------------------------------------------------------------------------------------------------

This is one of the most frequently asked SQL interview questions.

WHERE	HAVING

Filters individual rows	Filters groups

Executes before GROUP BY	Executes after GROUP BY

Cannot use Aggregate Functions directly	Commonly used with Aggregate Functions

Used before grouping	Used after grouping

-------------------------------------------------------------------------------------------------------------------------------

Example

WHERE

Show employees earning more than ₹50,000.

SELECT *
FROM Employees
WHERE Salary > 50000;

This filters individual employees.

-------------------------------------------------------------------------------------------------------------------------------

HAVING

Show departments whose average salary is greater than ₹50,000.

SELECT Department,
       AVG(Salary) AS Average_Salary
FROM Employees
GROUP BY Department
HAVING AVG(Salary) > 50000;

This filters departments, not employees.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

Remember this simple rule:

-------------------------------------------------------------------------------------------------------------------------------

WHERE  → Filters Rows

-------------------------------------------------------------------------------------------------------------------------------

HAVING → Filters Groups

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

📖 Business Question

-------------------------------------------------------------------------------------------------------------------------------

Query

       COUNT(*) AS Total_Employees
FROM Employees
GROUP BY Department
HAVING COUNT(*) > 1;

Explanation

-------------------------------------------------------------------------------------------------------------------------------

   ↓

   ↓

   ↓

1. WHERE + GROUP BY + HAVING

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The WHERE clause filters rows before grouping, GROUP BY creates groups, and HAVING filters those groups after aggregate calculations.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Show departments whose average salary is greater than ₹55,000, considering only employees with Experience ≥ 3 years.

Query

SELECT Department,
       AVG(Salary) AS Average_Salary
FROM Employees
WHERE Experience >= 3
GROUP BY Department
HAVING AVG(Salary) > 55000;

Output

Department	Average_Salary

IT	62500

-------------------------------------------------------------------------------------------------------------------------------

Explanation

Step 1 → Filter employees with Experience ≥ 3.

Step 2 → Group them by Department.

Step 3 → Calculate average salary.

Step 4 → Keep only departments with Average Salary > ₹55,000.

-------------------------------------------------------------------------------------------------------------------------------


💼 Business Insight

This pattern is used in almost every dashboard and reporting system.

-------------------------------------------------------------------------------------------------------------------------------

Examples:

Average sales of active stores

Revenue by city after filtering completed orders

Department-wise average salary of permanent employees

-------------------------------------------------------------------------------------------------------------------------------

2. HAVING with ORDER BY

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

ORDER BY sorts the grouped result after the HAVING clause filters the groups.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Display departments having more than one employee, sorted by employee count.

Query

SELECT Department,
       COUNT(*) AS Total_Employees
FROM Employees
GROUP BY Department
HAVING COUNT(*) > 1
ORDER BY Total_Employees DESC;

Output

Department	Total_Employees

HR	2

IT	2

Explanation

SQL first creates groups.

Then removes departments with only one employee.

Finally, it sorts the remaining groups.

-------------------------------------------------------------------------------------------------------------------------------

3. HAVING with Multiple Aggregate Functions
-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The HAVING clause can use more than one aggregate function in the same condition.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Show departments where:

Average Salary > ₹50,000
AND Employee Count > 1

Query

SELECT Department,
       COUNT(*) AS Total_Employees,
       AVG(Salary) AS Average_Salary
FROM Employees
GROUP BY Department
HAVING COUNT(*) > 1
   AND AVG(Salary) > 50000;

Output

Department	Total_Employees	Average_Salary

IT	2	62500

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

HAVING conditions can be combined using:

AND
OR
NOT

just like the WHERE clause.

8. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------


Using aggregate functions inside WHERE.

Wrong

WHERE COUNT(*) > 1;

Correct

HAVING COUNT(*) > 1;

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------


Using HAVING instead of WHERE for row filtering.

Wrong

HAVING Salary > 50000;

Correct

WHERE Salary > 50000;

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------


Thinking HAVING runs before GROUP BY.

Wrong.

GROUP BY executes first.

⚡ Performance Note

-------------------------------------------------------------------------------------------------------------------------------

Filtering rows with WHERE before grouping is generally more efficient than filtering only with HAVING, because fewer rows need to be grouped.

-------------------------------------------------------------------------------------------------------------------------------

9. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. Can WHERE and HAVING be used together?

Answer

Yes. WHERE filters rows before grouping, while HAVING filters groups after aggregation.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Which clause executes first: WHERE or HAVING?

Answer

WHERE executes before GROUP BY, while HAVING executes after aggregate calculations.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Can multiple conditions be used in HAVING?

Answer

Yes. HAVING supports logical operators such as AND, OR, and NOT.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Can ORDER BY be used with HAVING?

Answer

Yes. After HAVING filters the groups, ORDER BY can sort the final result.
