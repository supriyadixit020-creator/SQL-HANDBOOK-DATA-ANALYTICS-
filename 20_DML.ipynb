Section A : Introduction to DML

-------------------------------------------------------------------------------------------------------------------------------

1. What is DML?

📖 Definition (Interview Ready)

Data Manipulation Language (DML) is a category of SQL commands used to insert, update, delete, and retrieve data stored in database tables.



-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need DML?

A database is useful only if we can manage its data.

With DML we can:

Add new records.

Update existing records.

Delete unwanted records.

Retrieve stored information.

Without DML, tables would remain empty or outdated.

DML Commands

Command	Purpose

INSERT	Add new records

UPDATE	Modify existing records

DELETE	Remove records

MERGE	Insert, Update, or Delete in one operation (covered later)

💼 Business Scenario

An e-commerce company receives new customer registrations every day.

To keep the database updated, employees need to:

Add new customers.

Update customer information.

Remove inactive customer records.

These tasks are performed using DML commands.

The INSERT statement is used to add one or more new rows into a database table.

📖 Definition (Interview Ready)

The UPDATE statement modifies existing records in a table based on a specified condition.


-------------------------------------------------------------------------------------------------------------------------------

Syntax

⭐ Important Warning

📖 Definition (Interview Ready)

The DELETE statement removes one or more rows from a table based on a specified condition.

Best Practices

-------------------------------------------------------------------------------------------------------------------------------

✅ Always use a WHERE clause with UPDATE and DELETE unless you intentionally want to affect every row.

✅ Specify column names in INSERT statements whenever possible.

✅ Test UPDATE and DELETE using a SELECT statement first.

Example:

SELECT *

FROM Customers

WHERE CustomerID = 101;

✅ Take a backup before running UPDATE or DELETE on production databases.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is DML?

**Answer**

DML (Data Manipulation Language) is a set of SQL commands used to manipulate data stored in database tables, including INSERT, UPDATE, and DELETE.

Q2. What is the difference between DML and DDL?

**Answer**

DML commands (INSERT, UPDATE, DELETE) manipulate the data within tables, while DDL commands (CREATE, ALTER, DROP) define or modify the structure of database objects.

Q3. What is INSERT?

**Answer**

INSERT is a DML command used to add new rows of data into a table.

Q4. What is UPDATE?

**Answer**

UPDATE is a DML command used to modify existing data in one or more rows of a table.

Q5. What is DELETE?

**Answer**

DELETE is a DML command used to remove one or more existing rows from a table.

Q6. What happens if UPDATE is executed without a WHERE clause?

**Answer**

Without a WHERE clause, UPDATE modifies every row in the table, which can unintentionally overwrite all existing data.

Q7. What happens if DELETE is executed without a WHERE clause?

**Answer**

Without a WHERE clause, DELETE removes every row from the table, effectively emptying it while keeping the table structure intact.

Section A : MERGE Statement

-------------------------------------------------------------------------------------------------------------------------------

1. What is MERGE?

📖 Definition (Interview Ready)

The MERGE statement performs INSERT, UPDATE, and DELETE operations in a single SQL statement by comparing data between a source table and a target table.


-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need MERGE?

Suppose a company receives a daily customer file.

Some customers:

Already exist → Update

Are new → Insert

No longer exist → Delete (optional)

Instead of writing multiple queries, use a single MERGE statement.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Easy Visualization

Source Table

        │

        ▼

Compare

        │

 ┌──────┼──────┐

Update  Insert  Delete

Syntax

MERGE TargetTable AS T

USING SourceTable AS S

ON T.CustomerID = S.CustomerID

WHEN MATCHED THEN

UPDATE SET T.City = S.City

WHEN NOT MATCHED THEN

INSERT(CustomerID,CustomerName,City)

VALUES

(S.CustomerID,S.CustomerName,S.City);

💼 Business Scenario

Every night, a CRM system sends updated customer information.

The database should:

Update existing customers.
Insert new customers.

MERGE performs both tasks in one statement.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Pro Tip

MERGE is powerful but should be tested carefully in production. In SQL Server, some edge cases and concurrency issues have been reported over the years. Many organizations still prefer separate INSERT, UPDATE, and DELETE statements for critical systems. Knowing MERGE is important, but also know its limitations.

"What is UPSERT?"

Section C : Transactions

-------------------------------------------------------------------------------------------------------------------------------

2. What is a Transaction?

📖 Definition (Interview Ready)

A Transaction is a sequence of one or more SQL statements executed as a single logical unit of work. Either all operations succeed, or none of them are permanently applied.

-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need Transactions?

Imagine a bank transfer.

Rahul transfers ₹5,000 to Priya.

Steps:

Deduct ₹5,000 from Rahul.

Add ₹5,000 to Priya.

If the system crashes after Step 1,

Rahul loses money.

Priya doesn't receive it.

Transactions prevent this problem.

Section D : BEGIN TRANSACTION

-------------------------------------------------------------------------------------------------------------------------------

Definition

BEGIN TRANSACTION marks the starting point of a transaction.

Syntax

BEGIN TRANSACTION;
Example

UPDATE Accounts

WHERE AccountID = 101;

-------------------------------------------------------------------------------------------------------------------------------

Section E : COMMIT

-------------------------------------------------------------------------------------------------------------------------------

Definition

COMMIT permanently saves all changes made during the current transaction.

Syntax

COMMIT;
Example
BEGIN TRANSACTION;

UPDATE Accounts

SET Balance = Balance - 5000

WHERE AccountID = 101;

UPDATE Accounts

SET Balance = Balance + 5000

WHERE AccountID = 102;

COMMIT;
Result

✅ Money transferred successfully.

⭐ Interview Tip

Once a transaction is committed, the changes become permanent and cannot be undone using ROLLBACK.

-------------------------------------------------------------------------------------------------------------------------------

Section F : ROLLBACK

-------------------------------------------------------------------------------------------------------------------------------

Definition

ROLLBACK cancels all uncommitted changes made during the current transaction.

Syntax

ROLLBACK;
Example
BEGIN TRANSACTION;

UPDATE Accounts

SET Balance = Balance - 5000

WHERE AccountID = 101;

ROLLBACK;
Result

No changes are saved.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is DML?

**Answer**

DML (Data Manipulation Language) is a set of SQL commands used to manipulate data stored in database tables, including INSERT, UPDATE, and DELETE.

Q2. What is MERGE?

**Answer**

MERGE is a SQL command that combines INSERT, UPDATE, and DELETE operations into a single statement, allowing conditional synchronization between a source and a target table.

Q3. What is UPSERT?

**Answer**

UPSERT is the concept of inserting a new row if it does not exist, or updating the existing row if it does — commonly implemented in SQL using the MERGE statement.

Q4. What is a Transaction?

**Answer**

A Transaction is a sequence of one or more SQL operations executed as a single logical unit of work, which either completes entirely (COMMIT) or is fully undone (ROLLBACK).

Q5. What is COMMIT?

**Answer**

COMMIT is a command that permanently saves all the changes made during the current transaction to the database.

Q6. What is ROLLBACK?

**Answer**

ROLLBACK is a command that undoes all the changes made during the current transaction, reverting the database to its previous state.

Q7. What is the difference between COMMIT and ROLLBACK?

**Answer**

COMMIT permanently saves the changes made in a transaction, while ROLLBACK undoes those changes and restores the database to its state before the transaction began.

Q8. What are ACID properties?

**Answer**

ACID stands for Atomicity, Consistency, Isolation, and Durability — the four properties that guarantee reliable processing of database transactions.

Q9. Why are Transactions important?

**Answer**

Transactions ensure data integrity and consistency by making sure a group of related operations either all succeed together or all fail together, especially important in multi-step operations like fund transfers.

Q10. Give a real-world example of a Transaction.

**Answer**

A bank fund transfer is a classic example: money is debited from one account and credited to another; both operations must succeed together, or neither should be applied, to keep account balances consistent.
