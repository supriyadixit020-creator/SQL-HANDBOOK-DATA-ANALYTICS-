1. What is SQL?

SQL (Structured Query Language) is a standard programming language used to communicate with relational databases. It allows users to store, retrieve, update, delete, and manage data efficiently.

SQL is the standard language used to communicate with relational databases.

2. What is a Database?

A database is an organized collection of related data that is stored electronically and can be easily accessed, managed, and updated.

-------------------------------------------------------------------------------------------------------------------------------

3. What is a Table?

A table is a structured collection of related data organized into rows and columns within a database.

-------------------------------------------------------------------------------------------------------------------------------

4. What is a Row?


A row represents a single record or entry in a table. Each row contains complete information about one entity.

-------------------------------------------------------------------------------------------------------------------------------

5. what is a Column?

A column represents a specific attribute or field of the data stored in a table.

-------------------------------------------------------------------------------------------------------------------------------

6. What is DBMS?

DBMS (Database Management System) is software that allows users to create, store, retrieve, and manage data in a database

-------------------------------------------------------------------------------------------------------------------------------

7. What is RDBMS?

RDBMS (Relational Database Management System) is a type of DBMS that stores data in related tables and maintains relationships using keys such as Primary Key and Foreign Key.

9. What is DQL?

DQL (Data Query Language) is used to retrieve data from a database.

Example:

SELECT *
FROM Employees;

-------------------------------------------------------------------------------------------------------------------------------

10. What is DML?

DML (Data Manipulation Language) is used to insert, update, and delete data in a database.

Example:

INSERT INTO Employees
VALUES (101, 'John');

UPDATE Employees
SET Salary = 60000
WHERE Employee_ID = 101;

DELETE FROM Employees
WHERE Employee_ID = 101;

-------------------------------------------------------------------------------------------------------------------------------

11. What is DDL?

DDL (Data Definition Language) is used to create, modify, and delete database objects such as tables and databases.

Example:

CREATE TABLE Employees (
    Employee_ID INT,
    Name VARCHAR(50)
);

ALTER TABLE Employees
ADD Salary INT;

DROP TABLE Employees;

-------------------------------------------------------------------------------------------------------------------------------

12. What is DCL?

DCL (Data Control Language) is used to control user permissions and access to the database.

Example:

GRANT SELECT
ON Employees
TO User1;

REVOKE SELECT
ON Employees
FROM User1;

-------------------------------------------------------------------------------------------------------------------------------

13. What is TCL?

TCL (Transaction Control Language) is used to manage transactions in a database and ensure data consistency.

Example:

COMMIT;

ROLLBACK;

SAVEPOINT Save1;

Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Difference between SQL and MySQL?

**Answer**

SQL is a standard query language used to communicate with relational databases, while MySQL is a specific database management system (DBMS) that implements and uses SQL.

What is DBMS?

**Answer**

A DBMS (Database Management System) is software used to create, store, manage, and retrieve data in a database, e.g., MySQL, Oracle, SQL Server.

What is RDBMS?

**Answer**

An RDBMS (Relational DBMS) is a DBMS that stores data in structured tables with rows and columns, and enforces relationships between tables using keys.

Why SQL is called declarative language?

**Answer**

Because in SQL we specify *what* result we want, not *how* to get it — the database engine decides the execution steps internally.

What are SQL categories?

**Answer**

SQL commands are grouped into DDL (Data Definition Language), DML (Data Manipulation Language), DQL (Data Query Language), DCL (Data Control Language), and TCL (Transaction Control Language).

Is SQL case-sensitive?

**Answer**

SQL keywords are generally not case-sensitive (SELECT and select work the same), but table/column name case-sensitivity depends on the specific database system and its configuration.

Can SQL work without Primary Key?

**Answer**

Yes, a table can technically exist without a Primary Key, but it's not recommended since it can lead to duplicate rows and makes it harder to uniquely identify or relate records.

What are constraints?

**Answer**

Constraints are rules applied to table columns to maintain data accuracy and integrity, such as PRIMARY KEY, FOREIGN KEY, NOT NULL, UNIQUE, CHECK, and DEFAULT.

What is schema?

**Answer**

A schema is the logical structure or blueprint of a database — it defines how data is organized, including tables, columns, data types, and relationships.

What is metadata?

**Answer**

Metadata is "data about data" — information that describes the structure of a database, such as table names, column names, data types, and constraints.
