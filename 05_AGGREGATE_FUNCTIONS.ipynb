-------------------------------------------------------------------------------------------------------------------------------

Definition (Interview Ready)

Aggregate Functions are built-in SQL functions that perform calculations on multiple rows and return a single summarized value.

-------------------------------------------------------------------------------------------------------------------------------

Why Do We Use Aggregate Functions?

In Data Analytics, we often need summaries, not individual records.

-------------------------------------------------------------------------------------------------------------------------------

For example:

How many employees are there?

What is the total sales amount?

What is the average salary?

Which employee has the highest salary?

What is the minimum product price?

Instead of calculating these manually, SQL provides Aggregate Functions.

-------------------------------------------------------------------------------------------------------------------------------

What Are Aggregate Functions?

SQL provides the following commonly used aggregate functions:

Function	Purpose

COUNT()	Counts rows or values

SUM()	Calculates the total

AVG()	Calculates the average

MIN()	Finds the smallest value

MAX()	Finds the largest value

-------------------------------------------------------------------------------------------------------------------------------

📌 Interview Tip: These are the five most frequently used aggregate functions in SQL interviews.

-------------------------------------------------------------------------------------------------------------------------------

COUNT() Function

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The COUNT() function returns the number of rows or non-NULL values in a specified column.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT COUNT(column_name)
FROM table_name;

-------------------------------------------------------------------------------------------------------------------------------

Example 1 – Count All Employees

Business Question

-------------------------------------------------------------------------------------------------------------------------------

How many employees are there?

Query

SELECT COUNT(*)
FROM Employees;

Output

COUNT(*)

6

Explanation

COUNT(*) counts all rows, regardless of the values in each column.

Example 2 – Count Employee Names

SELECT COUNT(Name)
FROM Employees;

Output

COUNT(Name)

6

Explanation

Since the Name column contains no NULL values, the result is also 6.

📌 We'll learn the difference between COUNT(*) and COUNT(column_name) in Part 2.

-------------------------------------------------------------------------------------------------------------------------------

6. SUM() Function

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The SUM() function calculates the total of all numeric values in a column.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT SUM(column_name)
FROM table_name;

-------------------------------------------------------------------------------------------------------------------------------

Example
Business Question

-------------------------------------------------------------------------------------------------------------------------------

What is the total salary paid to all employees?

Query

SELECT SUM(Salary)
FROM Employees;

Output

SUM(Salary)

323000

Explanation

SQL adds:

45000 + 60000 + 55000 + 50000 + 65000 + 48000

= 323000

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

Finance teams often use SUM() to calculate:

Total Revenue
Total Profit
Total Sales
Total Expenses

-------------------------------------------------------------------------------------------------------------------------------

7. AVG() Function

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The AVG() function calculates the average value of a numeric column.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT AVG(column_name)
FROM table_name;

-------------------------------------------------------------------------------------------------------------------------------

Example

Business Question

-------------------------------------------------------------------------------------------------------------------------------

What is the average salary of employees?

Query

SELECT AVG(Salary)
FROM Employees;

Output

AVG(Salary)

53833.33

Explanation

Average = Total Salary ÷ Number of Employees

323000 ÷ 6 = 53833.33

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

Analysts commonly use AVG() to calculate:

Average Salary

Average Revenue

Average Customer Rating

Average Delivery Time

COUNT(*) vs COUNT(column_name)

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

COUNT(*) counts all rows in a table, whereas COUNT(column_name) counts only the non-NULL values in the specified column.

-------------------------------------------------------------------------------------------------------------------------------

Example 1 – COUNT(*)

Business Question

-------------------------------------------------------------------------------------------------------------------------------

How many employees are there?

Query

SELECT COUNT(*)
FROM Employees;

Output

COUNT(*)

6

Explanation

COUNT(*) counts every row, including rows where some columns contain NULL.

-------------------------------------------------------------------------------------------------------------------------------

Example 2 – COUNT(Email)

Business Question

-------------------------------------------------------------------------------------------------------------------------------

How many employees have an email address?

Query

SELECT COUNT(Email)
FROM Employees;

Output

COUNT(Email)

4

Explanation

COUNT(Email) ignores the NULL values.

Only 4 employees have email addresses.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

Suppose an interviewer asks:

"What is the difference between COUNT(*) and COUNT(column_name)?"

Answer:

COUNT(*) counts all rows in a table, while COUNT(column_name) counts only the non-NULL values in the specified column.

2. Aggregate Functions and NULL Values

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

Except for COUNT(*), aggregate functions ignore NULL values during calculations.

-------------------------------------------------------------------------------------------------------------------------------

Example

Suppose Salary values are:

Salary

45000

60000

NULL

50000

Query

SELECT AVG(Salary)
FROM Employees;

Output

51666.67

Explanation

SQL calculates:

(45000 + 60000 + 50000)

÷ 3

NOT

÷ 4

because the NULL value is ignored.

-------------------------------------------------------------------------------------------------------------------------------

Important Rule

-------------------------------------------------------------------------------------------------------------------------------

Function	Ignores NULL?

COUNT(*)	❌ No

COUNT(column)	✅ Yes

SUM()	✅ Yes

AVG()	✅ Yes

MIN()	✅ Yes

MAX()	✅ Yes

-------------------------------------------------------------------------------------------------------------------------------

📌 Remember: COUNT(*) counts rows, not values.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

In customer databases, many columns such as Phone Number, Email, or Address may contain NULL values.

Using COUNT(column_name) helps analysts determine how many customers actually have that information.

3. Aggregate Functions with WHERE

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The WHERE clause filters rows before the aggregate function performs its calculation.

-------------------------------------------------------------------------------------------------------------------------------

Example 1
Business Question

-------------------------------------------------------------------------------------------------------------------------------

What is the total salary of IT employees?

Query

SELECT SUM(Salary) AS Total_IT_Salary
FROM Employees
WHERE Department = 'IT';

Output

Total_IT_Salary

125000

Explanation

SQL performs the following steps:

FROM
   ↓
WHERE
   ↓
SUM()

Only IT employees are selected.

Then SQL adds their salaries.

-------------------------------------------------------------------------------------------------------------------------------

Example 2

Find the average salary of HR employees.

SELECT AVG(Salary) AS Average_HR_Salary
FROM Employees
WHERE Department = 'HR';

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

Managers rarely need reports for the entire company.

Instead, they ask:

Total sales in Delhi

Average salary of IT employees

Highest sales in 2025

These queries combine WHERE with aggregate functions.

4. Aggregate Functions with DISTINCT

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The DISTINCT keyword removes duplicate values before the aggregate function performs its calculation.

Example Table

Salary

45000

60000

60000

50000

-------------------------------------------------------------------------------------------------------------------------------

Example

Query

SELECT SUM(DISTINCT Salary)
FROM Employees;

Output

155000

Explanation

Duplicate salary 60000 is counted only once.

Calculation:

45000 + 60000 + 50000

= 155000

Another Example

SELECT COUNT(DISTINCT Department)
FROM Employees;

Output

Suppose departments are:

HR

IT

IT

Finance

Marketing

HR

Result

4

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

DISTINCT removes duplicate values before the aggregate function calculates the result.

-------------------------------------------------------------------------------------------------------------------------------

5. Logical Execution Order

-------------------------------------------------------------------------------------------------------------------------------

Consider the query:

SELECT AVG(Salary)
FROM Employees
WHERE Department = 'IT';

SQL processes the query logically as follows:

FROM
   ↓
WHERE
   ↓
AVG()
   ↓
SELECT

-------------------------------------------------------------------------------------------------------------------------------

Explanation

SQL reads the table.

Filters IT employees.

Calculates the average salary.

Displays the result.

⚠️ Common Interview Traps

-------------------------------------------------------------------------------------------------------------------------------

Trap 1
Wrong
SELECT COUNT(Name)
FROM Employees;

Many candidates say this counts all rows.

❌ Not always.

It counts only non-NULL values.

Correct Understanding
COUNT(*) → Counts all rows

COUNT(column) → Counts only non-NULL values

-------------------------------------------------------------------------------------------------------------------------------

Trap 2

Thinking SUM() works on text columns.

❌ Wrong

SELECT SUM(Name)
FROM Employees;

SUM() only works with numeric columns.

-------------------------------------------------------------------------------------------------------------------------------

Trap 3

Assuming AVG() includes NULL values.

❌ Wrong.

AVG() ignores NULL values.

-------------------------------------------------------------------------------------------------------------------------------

Trap 4

Using DISTINCT when values are already unique.

SELECT SUM(DISTINCT Salary)
FROM Employees;

If every salary is already unique, DISTINCT is unnecessary.

-------------------------------------------------------------------------------------------------------------------------------

Trap 5

Forgetting to use aliases.

Instead of:

SELECT SUM(Salary)
FROM Employees;

Prefer:

SELECT SUM(Salary) AS Total_Salary
FROM Employees;

This makes reports much easier to read.
