Section A : Introduction to DDL

-------------------------------------------------------------------------------------------------------------------------------

1. What is DDL?

📖 Definition (Interview Ready)

Data Definition Language (DDL) is a category of SQL commands used to create, modify, and remove database objects such as tables, databases, indexes, and views.



-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need DDL?

Before storing data, we must create the database structure.

DDL helps us:

Create tables

Modify tables

Delete tables

Reset tables

Design database schemas

Without DDL, a database cannot be built.

-------------------------------------------------------------------------------------------------------------------------------

DDL Commands

Command	Purpose

-------------------------------------------------------------------------------------------------------------------------------

CREATE	Create new database objects

ALTER	Modify existing objects

DROP	Remove objects permanently

TRUNCATE	Remove all rows while keeping the table structure

RENAME*	Rename database objects (syntax varies across database systems)

💼 Business Scenario

A new e-commerce company is being launched.

The database team needs to:

Create Customers table.

Create Products table.

Modify tables as business requirements change.

Remove old tables no longer needed.

These tasks are performed using DDL commands.

Section B : CREATE Statement

-------------------------------------------------------------------------------------------------------------------------------

2. What is CREATE?

📖 Definition (Interview Ready)

The CREATE statement is used to create new database objects such as tables, databases, views, indexes, and stored procedures.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

Create Database

CREATE DATABASE CompanyDB;

Create Table

CREATE TABLE Employees

(

EmployeeID INT PRIMARY KEY,

EmployeeName VARCHAR(100),

Department VARCHAR(50),

Salary DECIMAL(10,2)

);

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Requirement

Create an Employees table to store employee information.

Output

A new table named Employees is created.

Business Use Cases

CREATE is commonly used for:

New Applications

New Modules

Data Warehouses

Reporting Databases

-------------------------------------------------------------------------------------------------------------------------------

Section C : ALTER Statement

-------------------------------------------------------------------------------------------------------------------------------

3. What is ALTER?

📖 Definition (Interview Ready)

The ALTER statement modifies the structure of an existing database object without deleting it.



-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need ALTER?

Business requirements change over time.

Examples:

Add a new column.

Remove an old column.

Change a data type.

Rename a column (database-specific syntax).

Add Column

ALTER TABLE Employees

ADD Email VARCHAR(100);

Modify Data Type

ALTER TABLE Employees

ALTER COLUMN EmployeeName VARCHAR(150);

Drop Column

ALTER TABLE Employees

DROP COLUMN Email;

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Use Cases

Add GST Number

Add Mobile Number

Increase column size

Remove unused columns

-------------------------------------------------------------------------------------------------------------------------------

Section D : DROP Statement

-------------------------------------------------------------------------------------------------------------------------------

4. What is DROP?

📖 Definition (Interview Ready)

The DROP statement permanently removes a database object along with all of its data and structure.


-------------------------------------------------------------------------------------------------------------------------------

Syntax
DROP TABLE Employees;
Result

The table and all data are permanently removed.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Important Warning

Once a table is dropped,

Structure is deleted.

Data is deleted.

Constraints are deleted.

Indexes are deleted.

Recovery may not be possible without backups.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Use Cases

Remove obsolete tables.


Delete test databases.
Remove old backup tables.

-------------------------------------------------------------------------------------------------------------------------------

Section E : TRUNCATE Statement

-------------------------------------------------------------------------------------------------------------------------------

5. What is TRUNCATE?

📖 Definition (Interview Ready)

The TRUNCATE statement removes all rows from a table while keeping its structure, columns, constraints, and indexes.



-------------------------------------------------------------------------------------------------------------------------------

Syntax

TRUNCATE TABLE Employees;
Example

Before

EmployeeID	Name

101	Rahul

102	Priya

Run

TRUNCATE TABLE Employees;

After

EmployeeID	Name

(No Rows)	

The table still exists, but it is empty.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Use Cases

Clear staging tables.

Reset test environments.

Remove temporary data.

Prepare tables for fresh imports.

-------------------------------------------------------------------------------------------------------------------------------

Always review dependencies before using DROP.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is DDL?

**Answer**

DDL (Data Definition Language) is a set of SQL commands used to define, modify, and manage the structure of database objects such as tables, schemas, and indexes, including CREATE, ALTER, DROP, and TRUNCATE.

Q2. What is CREATE?

**Answer**

CREATE is a DDL command used to create new database objects such as tables, views, or databases.

Q3. What is ALTER?

**Answer**

ALTER is a DDL command used to modify the structure of an existing database object, such as adding, modifying, or dropping a column.

Q4. What is DROP?

**Answer**

DROP is a DDL command used to permanently delete an entire database object (such as a table) along with its structure and data.

Q5. What is TRUNCATE?

**Answer**

TRUNCATE is a DDL command used to quickly remove all rows from a table while keeping the table structure intact.

Q6. What is the difference between DROP and TRUNCATE?

**Answer**

DROP removes the entire table structure along with its data, while TRUNCATE removes only the data and keeps the table structure intact.

Q7. Which command removes only data but keeps the table?

**Answer**

TRUNCATE removes only the data while keeping the table structure intact.

Section G : Performance Tips

-------------------------------------------------------------------------------------------------------------------------------

✅ Tip 1

Use TRUNCATE instead of DELETE when removing all rows from a table and you don't need row-by-row deletion.

✅ Tip 2

Avoid frequent ALTER operations on large production tables during business hours.

✅ Tip 3

Always check dependencies before using DROP.

✅ Tip 4

Test DDL scripts in a development environment before applying them to production.

✅ Tip 5

Use meaningful and consistent naming conventions for tables and columns.
