🎯 Top Beginner SQL Interview Questions

Q1. What is SQL?
Answer

SQL (Structured Query Language) is a standard language used to create, manage, manipulate, and retrieve data stored in relational databases.

Q2. What is DBMS?
Answer

A Database Management System (DBMS) is software used to create, store, organize, and manage databases.

Examples:

MySQL
SQL Server
PostgreSQL
Oracle

Q3. What is RDBMS?
Answer

A Relational Database Management System (RDBMS) stores data in related tables using rows and columns and supports relationships through Primary Keys and Foreign Keys.

Q4. Difference between DBMS and RDBMS

DBMS	RDBMS

Stores Data	Stores Related Data

No Relationships	Supports Relationships

Less Structured	More Structured

Q5. What is a Table?

Answer

A table is a database object that stores related data in rows and columns.

Q6. What is a Row?
Answer

A row represents a single record in a table.

Q7. What is a Column?
Answer

A column represents a specific attribute or field of the data.

Q8. What is a Primary Key?
Answer

A Primary Key uniquely identifies each row in a table and does not allow duplicate or NULL values.

Q9. What is a Foreign Key?
Answer

A Foreign Key establishes a relationship between two tables by referencing the Primary Key of another table.

Q10. What is NULL?
Answer

NULL represents the absence of a value. It is different from zero, an empty string, or a space.

Q11. What is the difference between NULL and 0?
NULL	0
Missing Value	Numeric Value

Unknown	Known

Not Equal to Zero	Actual Number

Q12. What is the SELECT statement?
Answer

The SELECT statement is used to retrieve data from one or more tables.

Q13. What is the WHERE clause?
Answer

The WHERE clause filters records based on specified conditions.

Q14. What is ORDER BY?
Answer

ORDER BY sorts query results in ascending or descending order.

Q15. What is GROUP BY?
Answer

GROUP BY groups rows that have the same values into summary rows, usually with aggregate functions.

Q16. What is the difference between WHERE and HAVING?

Answer
WHERE	HAVING

Filters rows before grouping	Filters groups after grouping

Cannot use aggregate functions directly	Can use aggregate functions

Used before GROUP BY	Used after GROUP BY

Example

SELECT Department,

COUNT(*) AS TotalEmployees

FROM Employees

GROUP BY Department

HAVING COUNT(*) > 10;

⭐ Recruiter's Expectation

A good candidate should explain when each clause is executed, not just the syntax.

Q17. What is the difference between GROUP BY and ORDER BY?

Answer

GROUP BY	ORDER BY

Creates groups	Sorts results

Often used with aggregate functions	Can be used without aggregates

Changes result structure	Does not change result structure

Example

SELECT Department,

AVG(Salary) AS AvgSalary

FROM Employees

GROUP BY Department

ORDER BY AvgSalary DESC;

Q18. What is an INNER JOIN?
Answer

An INNER JOIN returns only the rows that have matching values in both tables.

Example

SELECT

C.CustomerName,

O.OrderID

FROM Customers C

INNER JOIN Orders O

ON C.CustomerID = O.CustomerID;


⭐ Recruiter's Expectation

Mention that unmatched rows are excluded.

Q19. Difference between INNER JOIN and LEFT JOIN

INNER JOIN	LEFT JOIN

Returns matching rows only	Returns all rows from the left table and matching rows from the right table

Excludes unmatched rows	Includes unmatched left-side rows

Example

Use LEFT JOIN when you want to find customers who have not placed any orders.

Q20. What is a CROSS JOIN?
Answer

A CROSS JOIN returns the Cartesian product of two tables, combining every row from the first table with every row from the second table.

Example

If Table A has 5 rows and Table B has 4 rows,

Result:

5 × 4 = 20 rows

Q21. What is a SELF JOIN?
Answer

A SELF JOIN joins a table to itself, usually using different aliases.

Real Example

Employees and Managers stored in the same table.

SELECT

E.EmployeeName,

M.EmployeeName AS Manager

FROM Employees E

LEFT JOIN Employees M

ON E.ManagerID = M.EmployeeID;

Q22. What is a Subquery?
Answer

A Subquery is a query written inside another SQL query.

Example

SELECT EmployeeName

FROM Employees

WHERE Salary >

(

SELECT AVG(Salary)

FROM Employees

);

⭐ Recruiter's Expectation

Mention that subqueries can return:

Single value

Multiple values

Entire result sets

Q23. What is a CTE?
Answer

A Common Table Expression (CTE) is a temporary named result set that simplifies complex SQL queries.

Example

WITH HighSalary AS

(

SELECT *

FROM Employees

WHERE Salary > 80000

)

SELECT *

FROM HighSalary;

Q24. CTE vs Subquery
CTE	Subquery

More readable	Less readable for complex logic

Can be referenced multiple times	Usually referenced once

Better for complex queries	Suitable for simpler queries

Q25. What is a Window Function?
Answer

A Window Function performs calculations across a set of rows without collapsing the result into a single row.

Common Window Functions

ROW_NUMBER()

RANK()

DENSE_RANK()

LAG()

LEAD()

Q26. Difference between RANK() and DENSE_RANK()

RANK()	DENSE_RANK()

Skips rank numbers after ties	Does not skip rank numbers

Example

Scores:

100
100
90

RANK()

1
1
3

DENSE_RANK()

1
1
2

Q27. What is ROW_NUMBER()?
Answer

ROW_NUMBER() assigns a unique sequential number to each row in the result set.

Example

SELECT

EmployeeName,

Salary,

ROW_NUMBER() OVER(ORDER BY Salary DESC) AS RowNum

FROM Employees;

Q28. Difference between ROW_NUMBER(), RANK(), and DENSE_RANK()

Function	Duplicate Values	Skips Numbers

ROW_NUMBER()	❌	❌

RANK()	✅	✅

DENSE_RANK()	✅	❌

Q29. What are Aggregate Functions?
Answer

Aggregate functions perform calculations on multiple rows and return a single result.

Examples:

COUNT()
SUM()
AVG()
MAX()
MIN()

Q30. What is CASE WHEN?
Answer

CASE WHEN is used to apply conditional logic within SQL queries.

Example

SELECT

EmployeeName,

Salary,

CASE

WHEN Salary >= 80000 THEN 'High'

WHEN Salary >= 50000 THEN 'Medium'

ELSE 'Low'

END AS SalaryCategory

FROM Employees;

Q31. What is the difference between PRIMARY KEY and UNIQUE?
Answer

PRIMARY KEY	UNIQUE

Uniquely identifies each row	Prevents duplicate values

Does not allow NULL values	SQL Server allows one NULL value

Only one PRIMARY KEY per table	Multiple UNIQUE constraints are allowed

⭐ Recruiter's Expectation

Mention that every PRIMARY KEY is UNIQUE, but every UNIQUE constraint is not a PRIMARY KEY.

Q32. What is the difference between DELETE, TRUNCATE, and DROP?

Answer

Feature	DELETE	TRUNCATE	DROP

Category	DML	DDL	DDL

Removes Data	✅	✅	✅

Removes Table Structure	❌	❌	✅

WHERE Clause	✅	❌	❌

Deletes Selected Rows	✅	❌	❌

Table Exists After Execution	✅	✅	❌

⭐ Recruiter's Expectation

Do not answer with only definitions.

Explain when each command should be used.

Q33. Difference between DML, DDL, DCL, and TCL

Answer

Language	Purpose	Examples

DML	Manipulate Data	INSERT, UPDATE, DELETE, MERGE

DDL	Define Database Objects	CREATE, ALTER, DROP, TRUNCATE

DCL	Control Permissions	GRANT, REVOKE

TCL	Manage Transactions	BEGIN TRANSACTION, COMMIT, ROLLBACK

Q34. What is a Transaction?
Answer

A Transaction is a sequence of SQL operations treated as a single unit of work. Either all operations succeed or none of them are permanently applied.

Real Example

Bank Transfer

Deduct Money

↓

Add Money

↓

COMMIT

OR

ROLLBACK

Q35. Difference between COMMIT and ROLLBACK

COMMIT	ROLLBACK

Saves changes permanently	Undoes uncommitted changes

Cannot be reversed using ROLLBACK	Restores previous state

Q36. What are ACID Properties?
Answer

Property	Meaning

Atomicity	All or Nothing

Consistency	Database remains valid

Isolation	Transactions do not interfere

Durability	Changes remain after COMMIT

-------------------------------------------------------------------------------------------------------------------------------

⭐ Easy Memory Trick

-------------------------------------------------------------------------------------------------------------------------------

A → Atomicity

C → Consistency

I → Isolation

D → Durability


Q37. What is a Constraint?
Answer

A Constraint is a rule applied to a table to maintain data accuracy, consistency, and integrity.

Common Constraints

PRIMARY KEY

FOREIGN KEY

UNIQUE

NOT NULL

DEFAULT

CHECK

Q38. What is Referential Integrity?

Answer

Referential Integrity ensures that every FOREIGN KEY value matches an existing PRIMARY KEY value in the parent table, preventing invalid relationships.

Q39. What is an Index?

Answer

An Index is a database object that improves the speed of data retrieval by allowing the database engine to locate rows more efficiently.

Benefits

Faster searching

Faster filtering

Faster JOIN operations

Drawback

Indexes improve read performance but may slow down INSERT, UPDATE, and DELETE operations because the index also needs to be maintained.

⭐ Recruiter's Expectation

Mention both advantages and disadvantages.

Q40. What is a View?

Answer

A View is a virtual table created from the result of a SQL query. It stores the query definition rather than the actual data.

Advantages

Simplifies complex queries.

Improves security by exposing only required columns.

Promotes query reuse.

Q41. What is a Temporary Table?

Answer

A Temporary Table stores temporary data that exists only for the duration of a session or scope.

Example

CREATE TABLE #Sales

(

SaleID INT,

Amount DECIMAL(10,2)

);

Q42. What is SQL Query Optimization?

Answer

SQL Query Optimization is the process of improving query performance while using fewer system resources.

Common Techniques

Create appropriate indexes.

Avoid SELECT * in production queries.

Filter early using WHERE.

Write efficient JOIN conditions.

Avoid unnecessary nested queries.

Q43. What is the SQL Execution Order?
Answer

SQL logically processes a query in this order:

FROM

↓

JOIN

↓

WHERE

↓

GROUP BY

↓

HAVING

↓

SELECT

↓

DISTINCT

↓

ORDER BY

↓

TOP / LIMIT

⭐ Recruiter's Expectation

Many candidates incorrectly answer using the written order instead of the execution order. Knowing the logical execution order demonstrates a deeper understanding of SQL.

Q44. What are Clustered and Non-Clustered Indexes?

Answer
Clustered Index	Non-Clustered Index

Determines the physical order of table data	Separate structure pointing to table data

One per table	Multiple allowed per table

Interview Tip

Understand the concept rather than memorizing internal implementation details.

Q45. What is Normalization?
Answer

Normalization is the process of organizing data into related tables to reduce redundancy and improve data integrity.

Common Normal Forms

First Normal Form (1NF)

Second Normal Form (2NF)

Third Normal Form (3NF)

Benefits

Reduces duplicate data.

Improves consistency.

Simplifies maintenance.

Q8. Calculate monthly sales.

YEAR(OrderDate) AS SalesYear,

MONTH(OrderDate) AS SalesMonth,

SUM(Amount) AS TotalSales

FROM Orders

YEAR(OrderDate),

MONTH(OrderDate);

Q9. Calculate running total.

OrderDate,

Amount,

SUM(Amount)

OVER

(

ORDER BY OrderDate

FROM Orders;

P.ProductName

FROM Products P

LEFT JOIN Sales S

ON P.ProductID = S.ProductID

WHERE S.ProductID IS NULL;
