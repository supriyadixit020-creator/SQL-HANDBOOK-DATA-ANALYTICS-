1. What is a Subquery?

📖 Definition (Interview Ready)

A Subquery is a query written inside another SQL query. The inner query executes first, and its result is used by the outer query.



-------------------------------------------------------------------------------------------------------------------------------

🎯 Easy Visualization

Outer Query

↓

Needs Some Information

↓

Runs Inner Query

↓

Gets Result

↓

Uses That Result

↓

Displays Final Output

-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Need Subqueries?

Suppose the manager asks:

"Find employees earning more than the average salary."

You don't know the average salary.

SQL first calculates it.

Then compares every employee.

This requires a Subquery.

Without Subquery

Step 1

SELECT AVG(Salary)
FROM Employees;

Suppose Output

60000

Step 2

SELECT *
FROM Employees
WHERE Salary > 60000;

This works, but it requires two separate queries.

With Subquery
SELECT *
FROM Employees
WHERE Salary >
(
    SELECT AVG(Salary)
    FROM Employees
);

One query.

Cleaner.

Professional.

💼 Business Insight

Subqueries help automate decision-making because SQL calculates intermediate values without requiring manual input.

4. ANY

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The ANY operator compares a value with each value returned by a subquery and returns TRUE if the condition is satisfied for at least one value.


-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

WHERE column_name > ANY
(
SELECT ...
)

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Find employees earning more than at least one HR employee.

Query

SELECT
EmployeeName,
Salary
FROM Employees
WHERE Salary > ANY
(
SELECT Salary
FROM Employees
WHERE DepartmentID = 1
);
HR Salaries

50000

55000

Result

Employees with salary greater than at least one HR salary.

Employee	Salary

Emma	70000

Sophia	80000

Michael	55000

Explanation

Since salaries are compared with each HR salary, the condition becomes TRUE if it is greater than any one of them.

-------------------------------------------------------------------------------------------------------------------------------

6. IN vs ANY vs ALL

-------------------------------------------------------------------------------------------------------------------------------

Operator	Meaning	Returns

IN	Match any value in a list	Equal values

ANY	Condition true for at least one value	One matching comparison is enough
ALL	Condition true for every value	Must satisfy all comparisons

-------------------------------------------------------------------------------------------------------------------------------

💼 Real Business Example

Business Question

Find employees working in departments where the average salary is above ₹60,000.

Query

SELECT

DepartmentID,
Salary
FROM Employees
WHERE DepartmentID IN
(
SELECT DepartmentID
FROM Employees
GROUP BY DepartmentID
HAVING AVG(Salary) > 60000
);

Step 1 (Subquery)

Calculate the average salary for each department.

Suppose the result is:

DepartmentID
2

Step 2 (Outer Query)

Return employees whose DepartmentID is 2.

1. What is a Correlated Subquery?

-------------------------------------------------------------------------------------------------------------------------------

A Correlated Subquery is a subquery that depends on the outer query. It executes once for every row processed by the outer query.

↓

↓

📌 Unlike a normal subquery, a correlated subquery runs once for every row.

3. What is a Nested Subquery?


📖 Definition (Interview Ready)

A Nested Subquery is a subquery placed inside another subquery, creating multiple levels of queries.


-------------------------------------------------------------------------------------------------------------------------------

🎯 Easy Visualization

-------------------------------------------------------------------------------------------------------------------------------


Outer Query


↓

Subquery

↓

Another Subquery

↓

Final Result

Example

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------

Find employees working in the department with the highest average salary.

Query

SELECT

    EmployeeName,

    DepartmentID,

    Salary

FROM Employees

WHERE DepartmentID =

(

    SELECT TOP 1 DepartmentID

    FROM

    (
        SELECT

            DepartmentID,

            AVG(Salary) AS Average_Salary

        FROM Employees

        GROUP BY DepartmentID

    ) AS Dept_Avg

    ORDER BY Average_Salary DESC

);

Employee	Department	Salary

Emma	2	70000

Sophia	2	80000

Explanation

Inner Query

Calculates average salary for each department.

↓

Department 2 has the highest average.

↓

Outer Query returns employees from Department 2.

Deeply nested subqueries can become difficult to read and maintain. For complex logic, a CTE (Common Table Expression) often provides a cleaner and more readable solution—which you'll learn in the next chapter.

Answer

A Derived Table is a temporary table created using a subquery inside the FROM clause.

A Nested Subquery is a subquery inside another subquery.

SQL requires every Derived Table to have an alias so the outer query can reference it.

💼 Real Business Scenarios

------------------------------------------------------------------------------------------------------------------------------

🏢 Scenario 1 – HR

Find employees earning above the company average.

Solution
SELECT *
FROM Employees
WHERE Salary >
(
SELECT AVG(Salary)
FROM Employees
);

------------------------------------------------------------------------------------------------------------------------------

🛒 Scenario 2 – E-commerce

Find customers who placed at least one order.

Solution
SELECT *
FROM Customers c
WHERE EXISTS
(
SELECT *
FROM Orders o
WHERE o.CustomerID = c.CustomerID
);

------------------------------------------------------------------------------------------------------------------------------

🏦 Scenario 3 – Banking

Find branches with no transactions.

Solution
SELECT *
FROM Branches b
WHERE NOT EXISTS
(
SELECT *
FROM Transactions t
WHERE t.BranchID = b.BranchID
);

------------------------------------------------------------------------------------------------------------------------------

📦 Scenario 4 – Inventory

Find products in categories with average price above ₹500.

Solution
SELECT *
FROM Products
WHERE CategoryID IN
(
SELECT CategoryID
FROM Products
GROUP BY CategoryID
HAVING AVG(Price) > 500
);

📊 Scenario 5 – Dashboard

Create a department salary summary.

Solution
SELECT *
FROM
(
SELECT
DepartmentID,
AVG(Salary) AS Avg_Salary
FROM Employees
GROUP BY DepartmentID
) AS DepartmentSummary;

7. Common Interview Traps

------------------------------------------------------------------------------------------------------------------------------

❌ Trap 1

Using = with a Multi-Row Subquery.

Wrong

WHERE DepartmentID =
(
SELECT DepartmentID
FROM Departments
);

Correct

WHERE DepartmentID IN
(
SELECT DepartmentID
FROM Departments
);

------------------------------------------------------------------------------------------------------------------------------

❌ Trap 2

-------------------------------------------------------------------------------------------------------------------------------

Forgetting to give a Derived Table an alias.

Wrong

FROM
(
SELECT ...
)

Correct

FROM
(
SELECT ...
)
AS DepartmentSummary

------------------------------------------------------------------------------------------------------------------------------

❌ Trap 3

-------------------------------------------------------------------------------------------------------------------------------

Confusing IN and EXISTS.

IN → Compare values.
EXISTS → Check whether related rows exist.

------------------------------------------------------------------------------------------------------------------------------

❌ Trap 4

-------------------------------------------------------------------------------------------------------------------------------

Assuming Correlated Subqueries execute once.

They execute once for every row processed by the outer query.

------------------------------------------------------------------------------------------------------------------------------

❌ Trap 5

-------------------------------------------------------------------------------------------------------------------------------

Writing deeply nested Subqueries without formatting.

Always indent every nested level.
