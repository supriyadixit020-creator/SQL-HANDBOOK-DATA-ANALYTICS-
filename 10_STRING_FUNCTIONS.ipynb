Section A : Text Case Functions

-------------------------------------------------------------------------------------------------------------------------------

4. UPPER()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The UPPER() function converts all characters in a string to uppercase letters.


Syntax

UPPER(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display all employee names in uppercase.

Query

SELECT
    EmployeeName,
    UPPER(EmployeeName) AS Upper_Name
FROM Employees;

Output

EmployeeName	Upper_Name

john	JOHN

EMMA	EMMA

DaViD	DAVID

sophia	SOPHIA

Explanation

Every alphabetic character is converted to uppercase.

Numbers and symbols remain unchanged.

💼 Business Use Cases

Standardizing names

Standardizing city names

Product code formatting

Report formatting

-------------------------------------------------------------------------------------------------------------------------------

5. LOWER()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The LOWER() function converts all characters in a string to lowercase letters.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

LOWER(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display all employee names in lowercase.

Query

SELECT
    EmployeeName,
    LOWER(EmployeeName) AS Lower_Name
FROM Employees;

Output

EmployeeName	Lower_Name

john	john

EMMA	emma

DaViD	david

sophia	sophia

Explanation

All alphabetic characters become lowercase.

Spaces remain unchanged.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

Email addresses are often converted to lowercase before comparison because email systems usually treat uppercase and lowercase letters as equivalent.

-------------------------------------------------------------------------------------------------------------------------------

6. Combining UPPER() and LOWER()

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display employee names in uppercase and lowercase together.

Query

SELECT
    EmployeeName,
    UPPER(EmployeeName) AS Upper_Name,
    LOWER(EmployeeName) AS Lower_Name
FROM Employees;

Output

EmployeeName	Upper_Name	Lower_Name

john	JOHN	john

EMMA	EMMA	emma

DaViD	DAVID	david

sophia	SOPHIA	sophia

-------------------------------------------------------------------------------------------------------------------------------

7. Practical Example

-------------------------------------------------------------------------------------------------------------------------------

Suppose users enter:

john
John
JOHN
JoHn

Convert everything to one format.

SELECT
    UPPER(EmployeeName) AS Standard_Name
FROM Employees;

Now all records become:

JOHN

This improves consistency and simplifies searching and reporting.

9. Best Practices

-------------------------------------------------------------------------------------------------------------------------------

✅ Store original data whenever possible.

✅ Use String Functions while querying or reporting rather than permanently changing source data unless required.

✅ Use meaningful aliases.

Example:

UPPER(EmployeeName) AS Upper_Name

-------------------------------------------------------------------------------------------------------------------------------

10. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Assuming UPPER() removes spaces.

Wrong.

It only changes letter case.

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Using LOWER() to clean spaces.

Use TRIM() instead.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Updating source data unnecessarily.

Prefer formatting during reporting unless the business specifically requires permanent changes.

-------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Note

Applying functions to indexed text columns in filtering conditions can reduce index usage in some databases. Use them thoughtfully, especially on very large tables.

-------------------------------------------------------------------------------------------------------------------------------

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What are String Functions?

Answer

String Functions are built-in SQL functions used to manipulate, clean, and format text data.

Q2. What does UPPER() do?

Answer

It converts all alphabetic characters in a string to uppercase.

Q3. What does LOWER() do?

Answer

It converts all alphabetic characters in a string to lowercase.

Q4. Do UPPER() and LOWER() remove spaces?

Answer

No. They only change the letter case. Spaces remain unchanged.

-------------------------------------------------------------------------------------------------------------------------------

📚 Quick Revision

-------------------------------------------------------------------------------------------------------------------------------

Section A : Length Functions

-------------------------------------------------------------------------------------------------------------------------------

1. LEN()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The LEN() function returns the number of characters in a string, excluding trailing spaces in SQL Server.

-------------------------------------------------------------------------------------------------------------------------------

📌 Note: LEN() is used in SQL Server.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

LEN(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Find the length of each employee's name.

Query

SELECT
    EmployeeName,
    LEN(EmployeeName) AS Name_Length
FROM Employees;

Output

EmployeeName	Name_Length

John	4

Emma	5

David	5

Sophia	7

Michael	7

Explanation

LEN() counts:

Letters

Numbers

Symbols

Leading spaces

But trailing spaces are ignored in SQL Server.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

LEN() is commonly used for:

Password validation

Phone number validation

Customer ID validation

Product code verification

-------------------------------------------------------------------------------------------------------------------------------

2. LENGTH()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

LENGTH() returns the total number of characters in a string. It is commonly used in MySQL and PostgreSQL.

📌 SQL Server uses LEN() instead of LENGTH().

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

LENGTH(column_name)

SQL Server vs MySQL

SQL Server	MySQL / PostgreSQL

LEN()	LENGTH()

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

Section B : Trimming Functions

-------------------------------------------------------------------------------------------------------------------------------

3. TRIM()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The TRIM() function removes leading and trailing spaces from a string.


-------------------------------------------------------------------------------------------------------------------------------

Syntax

TRIM(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display employee names after removing extra spaces.

Query

SELECT
    EmployeeName,
    TRIM(EmployeeName) AS Clean_Name
FROM Employees;

Output

EmployeeName	Clean_Name

John	John

Emma	Emma

David	David

Sophia	Sophia

Michael	Michael

Explanation

TRIM() removes:

✔ Leading spaces

✔ Trailing spaces

But does not remove spaces between words.

Example:

John   Smith

remains

John   Smith

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

TRIM is heavily used during:

CSV imports

Excel imports

Customer data cleaning

ETL processes

-------------------------------------------------------------------------------------------------------------------------------

4. LTRIM()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The LTRIM() function removes spaces only from the left (beginning) side of a string.


-------------------------------------------------------------------------------------------------------------------------------

Syntax

LTRIM(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Query

SELECT
    EmployeeName,
    LTRIM(EmployeeName) AS Left_Trimmed
FROM Employees;

Output

EmployeeName	Left_Trimmed

John	John

Emma	Emma

David	David

Sophia	Sophia

Michael	Michael

Explanation

Leading spaces are removed.

Trailing spaces remain unchanged.


-------------------------------------------------------------------------------------------------------------------------------

5. RTRIM()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The RTRIM() function removes spaces only from the right (end) side of a string.

-------------------------------------------------------------------------------------------------------------------------------
Syntax

RTRIM(column_name)

-------------------------------------------------------------------------------------------------------------------------------

Query

SELECT
    EmployeeName,
    RTRIM(EmployeeName) AS Right_Trimmed
FROM Employees;

Output

EmployeeName	Right_Trimmed

John	John

Emma	Emma

David	David

Sophia	Sophia

Michael	Michael

Explanation


Trailing spaces are removed.

Leading spaces remain unchanged.

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

Function	Removes Left Spaces	Removes Right Spaces

9. Best Practices

-------------------------------------------------------------------------------------------------------------------------------

✅ Clean imported text before analysis.

✅ Use TRIM() instead of combining LTRIM() and RTRIM() whenever your SQL version supports it.

✅ Validate important fields using LEN().

-------------------------------------------------------------------------------------------------------------------------------

10. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Thinking LEN() removes spaces.

Wrong.

It only counts characters.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Using LTRIM() when both leading and trailing spaces must be removed.

Use TRIM() instead.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Expecting TRIM() to remove spaces between words.

Wrong.

TRIM() removes spaces only from the beginning and end of the string.

-------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Note

String functions are generally fast, but repeatedly applying them to very large text columns in filtering conditions may reduce performance. Clean data during ETL whenever possible instead of repeatedly trimming the same values in reports.

-------------------------------------------------------------------------------------------------------------------------------

Answer

Q2. What is the difference between LEN() and LENGTH()?

Q3. What does TRIM() do?

Answer

TRIM() removes leading and trailing spaces from a string.

Q4. What is the difference between LTRIM() and RTRIM()?

Answer

LTRIM() removes spaces from the beginning of a string, while RTRIM() removes spaces from the end.

1. LEFT()

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The LEFT() function returns a specified number of characters from the beginning (left side) of a string.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

LEFT(column_name, number_of_characters)

-------------------------------------------------------------------------------------------------------------------------------

Emma Brown	Emma

David Miller	Davi

Sophia Jones	Soph

Explanation

LEFT(EmployeeName, 4)

Start from the left and return the first 4 characters.

Section C : SUBSTRING()

-------------------------------------------------------------------------------------------------------------------------------
4. SUBSTRING()

📖 Definition (Interview Ready)

The SUBSTRING() function extracts a specified number of characters starting from a given position within a string.

-------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

SUBSTRING(column_name, start_position, length)

-------------------------------------------------------------------------------------------------------------------------------

Parameters

Parameter	Meaning

column_name	Text column

start_position	Starting character position (starts from 1)

length	Number of characters to return

Business Question


Extract the first name from employee names.

Query

SELECT
    EmployeeName,
    SUBSTRING(EmployeeName,1,4) AS First_Name
FROM Employees;

Output

EmployeeName	First_Name

John Smith	John

Emma Brown	Emma

David Miller	Davi

Sophia Jones	Soph

Explanation

John Smith

123456789

Start Position = 1

Length = 4

Result

John

3. REPLACE()

-------------------------------------------------------------------------------------------------------------------------------

The REPLACE() function replaces all occurrences of a specified substring with another substring.

Syntax

REPLACE(column_name,'old_text','new_text')

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

Query

    Email,
    REPLACE(Email,'gmail.com','company.com') AS Official_Email

john@gmail.com	john@company.com

emma@yahoo.com	emma@yahoo.com

david@hotmail.com	david@hotmail.com

sophia@gmail.com	sophia@company.com

Explanation

Only matching text is replaced.

Other email domains remain unchanged.

LENGTH()	Returns string length (MySQL/PostgreSQL)	LENGTH('David') = 5

TRIM()	Removes leading & trailing spaces	' John ' → John

LTRIM()	Removes leading spaces	' John' → John

RTRIM()	Removes trailing spaces	'John ' → John

LEFT()	Extracts characters from the left	LEFT('David',2) → Da

RIGHT()	Extracts characters from the right	RIGHT('David',2) → id

SUBSTRING()	Extracts text from any position	SUBSTRING('David',2,3) → avi

CONCAT()	Combines strings	John Smith

REPLACE()	Replaces text	gmail.com → company.com

REVERSE()	Reverses characters	John → nhoJ

3. Common Interview Traps

-------------------------------------------------------------------------------------------------------------------------------

❌ Trap 1

Confusing LEN() with LENGTH().

Remember:

Database	Function
SQL Server	LEN()
MySQL	LENGTH()
PostgreSQL	LENGTH()


❌ Trap 2

Thinking TRIM() removes spaces between words.

Wrong.

It removes only leading and trailing spaces.

❌ Trap 3

Using LEFT() instead of SUBSTRING().

If extraction starts from the middle, use SUBSTRING().

❌ Trap 4

Forgetting a space in CONCAT().

Wrong

CONCAT(FirstName,LastName)

Output

JohnSmith

Correct

CONCAT(FirstName,' ',LastName)

Output

John Smith

❌ Trap 5

Expecting REPLACE() to modify stored data automatically.

It only changes the query result unless used in an UPDATE statement.
