1. CASE with WHERE

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The CASE statement is commonly used in the SELECT clause to create calculated values, while the WHERE clause filters rows before they are returned.

📌 Important: CASE does not replace WHERE. They work together.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display employees earning more than ₹50,000 and classify their salary.

Query

SELECT
    EmployeeName,
    Salary,
    CASE
        WHEN Salary >= 70000 THEN 'High Salary'
        ELSE 'Medium Salary'
    END AS Salary_Category
FROM Employees
WHERE Salary > 50000;

Output

EmployeeName	Salary	Salary_Category

Emma	75000	High Salary

David	60000	Medium Salary

Michael	90000	High Salary

Olivia	55000	Medium Salary

Explanation

Step 1 → WHERE removes employees earning ₹50,000 or less.

Step 2 → CASE categorizes the remaining employees.

💼 Business Insight

This pattern is used in payroll and HR reports where only eligible employees are analyzed.

-------------------------------------------------------------------------------------------------------------------------------

2. CASE with ORDER BY

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

A CASE statement can be used inside ORDER BY to create custom sorting logic instead of normal alphabetical or numerical sorting.

Business Question

Show employees in this order:

IT

HR

Finance

Marketing

Query

SELECT
    EmployeeName,
    Department
FROM Employees
ORDER BY
CASE
    WHEN Department = 'IT' THEN 1
    WHEN Department = 'HR' THEN 2
    WHEN Department = 'Finance' THEN 3
    ELSE 4
END;

Output

EmployeeName	Department

Emma	IT

Michael	IT

John	HR

Olivia	HR

David	Finance

Sophia	Marketing

Explanation

Instead of sorting alphabetically, SQL sorts using the numbers returned by the CASE expression.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

CASE inside ORDER BY is often used to create custom business sorting.

Example:

High Priority

Medium Priority

Low Priority

instead of alphabetical order.

3. CASE with GROUP BY
📖 Definition

-------------------------------------------------------------------------------------------------------------------------------

CASE can create custom categories, and GROUP BY can then summarize data based on those categories.

Business Question

Count employees in each salary category.

Query

SELECT
    CASE
        WHEN Salary >= 70000 THEN 'High Salary'
        WHEN Salary >= 50000 THEN 'Medium Salary'
        ELSE 'Low Salary'
    END AS Salary_Category,
    COUNT(*) AS Total_Employees
FROM Employees
GROUP BY
CASE
    WHEN Salary >= 70000 THEN 'High Salary'
    WHEN Salary >= 50000 THEN 'Medium Salary'
    ELSE 'Low Salary'
END;

Output

Salary_Category	Total_Employees

High Salary	2

Medium Salary	2

Low Salary	2


Explanation

The CASE statement creates salary groups, and GROUP BY counts employees in each group.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

This technique is commonly used to create:

Customer Segments

Salary Bands

Risk Levels

Product Categories

4. CASE with Aggregate Functions
Business Question

-------------------------------------------------------------------------------------------------------------------------------

Display the number of employees earning at least ₹60,000.

Query
SELECT
    SUM(
        CASE
            WHEN Salary >= 60000 THEN 1
            ELSE 0
        END
    ) AS High_Salary_Employees
FROM Employees;
Output
High_Salary_Employees

3

Explanation


For every employee:

Salary ≥ 60000 → return 1
Otherwise → return 0

SUM() adds the values to produce the final count.

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

This technique is called Conditional Aggregation and is one of the most commonly asked SQL interview concepts.

-------------------------------------------------------------------------------------------------------------------------------

5. CASE with AVG()
Business Question

-------------------------------------------------------------------------------------------------------------------------------


Find the average salary of experienced employees (Experience ≥ 5 years).

Query

SELECT

    AVG(

        CASE

            WHEN Experience >= 5 THEN Salary

        END

    ) AS Average_Experienced_Salary


FROM Employees;

Explanation


Employees with less than 5 years of experience return NULL, and AVG() ignores NULL values.

1. CASE with Multiple Conditions

-------------------------------------------------------------------------------------------------------------------------------
📖 Definition (Interview Ready)

A CASE statement can evaluate multiple conditions by using logical operators such as AND and OR inside each WHEN clause.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------

Classify employees based on Salary and Experience.

Rules:

Salary ≥ 70000 AND Experience ≥ 5 → Senior Employee
Salary ≥ 50000 → Mid-Level Employee
Otherwise → Junior Employee

Query

SELECT
    EmployeeName,
    Salary,
    Experience,
    CASE
        WHEN Salary >= 70000 AND Experience >= 5
            THEN 'Senior Employee'
        WHEN Salary >= 50000
            THEN 'Mid-Level Employee'
        ELSE 'Junior Employee'
    END AS Employee_Level
FROM Employees;

Output

EmployeeName	Employee_Level

John	Junior Employee

Emma	Senior Employee

David	Mid-Level Employee

Sophia	Junior Employee

Michael	Senior Employee

Olivia	Mid-Level Employee

Explanation

Emma:

✔ Salary ≥ 70000

✔ Experience ≥ 5

Result:

Senior Employee

3. CASE with INNER JOIN

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Display:

Employee Name
Department Name
Salary Category

Query

SELECT
    e.EmployeeName,
    d.DepartmentName,
    CASE
        WHEN e.Salary >= 70000 THEN 'High Salary'
        WHEN e.Salary >= 50000 THEN 'Medium Salary'
        ELSE 'Low Salary'
    END AS Salary_Category
FROM Employees AS e
INNER JOIN Departments AS d
    ON e.DepartmentID = d.DepartmentID;

Output

EmployeeName	DepartmentName	Salary_Category

John	HR	Low Salary

Emma	IT	High Salary

David	IT	Medium Salary

Sophia	Finance	Low Salary

Michael	IT	High Salary

Olivia	HR	Medium Salary

Explanation

The JOIN retrieves the department name, while CASE creates a readable salary category.

4. CASE with Aggregate Functions

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Count:

High Salary Employees

Medium Salary Employees

Low Salary Employees

Query

SELECT
    SUM(CASE WHEN Salary >= 70000 THEN 1 ELSE 0 END) AS High_Salary,
    SUM(CASE WHEN Salary BETWEEN 50000 AND 69999 THEN 1 ELSE 0 END) AS Medium_Salary,
    SUM(CASE WHEN Salary < 50000 THEN 1 ELSE 0 END) AS Low_Salary
FROM Employees;

Output

High_Salary	Medium_Salary	Low_Salary

2	2	2

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

This is one of the most common examples of Conditional Aggregation.

Many SQL interviews ask candidates to count different categories using a single query.

8. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Using overlapping conditions.

Wrong

WHEN Salary > 50000
WHEN Salary > 70000

The second condition will never be reached for salaries above 70000.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Making CASE unnecessarily complex.

If the logic becomes difficult to read, consider using a CTE or subquery.

-------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Not testing all possible outcomes.

Always verify that every row falls into the correct category.

-------------------------------------------------------------------------------------------------------------------------------

⚡ Performance Note

CASE itself is generally efficient. Performance issues usually come from complex expressions or repeated calculations. If the same CASE logic is reused several times, calculate it once in a CTE or subquery.
