Section A : Introduction to Views

-------------------------------------------------------------------------------------------------------------------------------

1. What is a View?

📖 Definition (Interview Ready)

A View is a virtual table in SQL that is created using the result of a SELECT statement. Unlike a regular table, a View does not store data itself. Instead, it stores the SQL query and displays data from the underlying table(s) whenever it is accessed.

-------------------------------------------------------------------------------------------------------------------------------


Why Do We Need Views?

As databases grow, SQL queries become longer and more complex.

Instead of writing the same query repeatedly, we can save it as a View and reuse it whenever needed.

Views help:

Reduce repetitive SQL code.

Simplify complex queries.

Improve readability.

Control data access.

Make reporting easier.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Scenario

Imagine an HR department that needs this report every day:

Employee Name

Department

Salary

Manager Name

The query requires multiple JOINs.

Instead of writing the same query every morning, the SQL developer creates a View.

Now HR only needs:

SELECT *
FROM Employee_Report;

This saves time and reduces the chance of errors.

Section B : Characteristics of a View

-------------------------------------------------------------------------------------------------------------------------------

A View has the following characteristics:

✅ It behaves like a table.

✅ It is created using a SELECT statement.

✅ It does not store data separately.

✅ It always retrieves the latest data from the underlying table(s).

✅ It can be used in SELECT statements just like a table.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Interview Tip

A View stores the query definition, not the actual data.

Whenever you query a View, SQL executes the underlying query and returns the latest results.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a View?

**Answer**

A View is a virtual table based on the result of a stored SQL query; it does not store data itself but displays data from the underlying base table(s).

Q2. Does a View store data?

**Answer**

No. A standard View does not store data physically; it stores only the SQL query definition and retrieves live data from the base tables each time it is queried.

Q3. Why are Views used?

**Answer**

Views are used to simplify complex queries, provide a layer of security by restricting access to specific columns/rows, and present data in a reusable, consistent format.

Q4. What are the advantages of using Views?

**Answer**

Views simplify complex queries, improve security by hiding sensitive columns, provide reusability, and present a consistent, simplified interface to the underlying data.

Q5. Can a View be queried like a table?

**Answer**

Yes. A View can be queried using SELECT statements just like a regular table.

Section A : ALTER VIEW

-------------------------------------------------------------------------------------------------------------------------------

1. What is ALTER VIEW?

📖 Definition (Interview Ready)

The ALTER VIEW statement is used to modify the definition of an existing View without deleting and recreating it.

-------------------------------------------------------------------------------------------------------------------------------


Why Do We Need ALTER VIEW?

Business requirements change over time.

Instead of deleting a View and creating it again, we can simply modify it.

Syntax

ALTER VIEW View_Name AS

SELECT Column1,
       Column2
FROM Table_Name
WHERE Condition;

Section B : DROP VIEW

-------------------------------------------------------------------------------------------------------------------------------

2. What is DROP VIEW?

📖 Definition (Interview Ready)

The DROP VIEW statement permanently removes a View from the database. It does not delete the underlying tables or their data.

-------------------------------------------------------------------------------------------------------------------------------


Syntax

DROP VIEW View_Name;
Example
DROP VIEW HighSalaryEmployees;
Important Note

After dropping the View,

This query will produce an error.

SELECT *
FROM HighSalaryEmployees;

-------------------------------------------------------------------------------------------------------------------------------

⭐ Interview Tip

-------------------------------------------------------------------------------------------------------------------------------

Dropping a View does not affect the original tables.

Only the View definition is removed.

Section C : Can We Update Data Through a View?

-------------------------------------------------------------------------------------------------------------------------------

Business Question

Can we perform

INSERT
UPDATE
DELETE

using a View?

Answer

✅ Sometimes.

It depends on how the View was created.

Example

Suppose we have:

CREATE VIEW EmployeeView AS

SELECT
EmployeeID,
EmployeeName,
Salary

FROM Employees;

Updating through the View

UPDATE EmployeeView

SET Salary = 65000

WHERE EmployeeID = 101;

If the View is updatable,

The Employees table is also updated.

Important Concept

A View does not have its own copy of the data.

Changes made through an updatable View affect the underlying table.

-------------------------------------------------------------------------------------------------------------------------------

Section D : Updatable Views

-------------------------------------------------------------------------------------------------------------------------------

What is an Updatable View?

📖 Definition

An Updatable View allows INSERT, UPDATE, and DELETE operations because it directly references data from a single base table without complex SQL operations.

-------------------------------------------------------------------------------------------------------------------------------

Characteristics

Generally, an Updatable View:

✅ Uses one table.

✅ Does not contain aggregate functions.

✅ Does not contain GROUP BY.

✅ Does not contain DISTINCT.

✅ Does not contain UNION.

Example
CREATE VIEW EmployeeView AS

SELECT
EmployeeID,
EmployeeName,
Salary

FROM Employees;

This View is generally updatable.

-------------------------------------------------------------------------------------------------------------------------------

Section E : Non-Updatable Views

-------------------------------------------------------------------------------------------------------------------------------

What is a Non-Updatable View?

📖 Definition

A Non-Updatable View does not allow direct INSERT, UPDATE, or DELETE operations because its definition includes operations that prevent SQL from determining how changes should be applied to the underlying data.

-------------------------------------------------------------------------------------------------------------------------------

Common Reasons

A View is generally not updatable if it contains:

❌ GROUP BY

❌ DISTINCT

❌ Aggregate Functions

❌ UNION

❌ INTERSECT

❌ EXCEPT

Example
CREATE VIEW DepartmentSummary AS

SELECT
Department,
AVG(Salary) AS AverageSalary

FROM Employees

GROUP BY Department;

Trying to update this View will result in an error.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is ALTER VIEW?

**Answer**

ALTER VIEW is used to modify the definition of an existing View without dropping and recreating it.

Q2. What is DROP VIEW?

**Answer**

DROP VIEW is used to permanently delete a View definition from the database.

Q3. Does DROP VIEW delete data from the base table?

**Answer**

No. DROP VIEW only removes the View definition; the underlying base table and its data remain unaffected.

Q4. What is an Updatable View?

**Answer**

An Updatable View is a View that allows INSERT, UPDATE, or DELETE operations to be performed on it, which are then reflected in the underlying base table.

Q5. What makes a View Non-Updatable?

**Answer**

A View becomes non-updatable when it contains JOINs, GROUP BY, aggregate functions, DISTINCT, or set operators, since SQL cannot unambiguously map the change back to a single base table row.

Q6. Can a View contain JOINs?

**Answer**

Yes, a View can contain JOINs, but such Views are generally not updatable.

Q7. Can a View contain Aggregate Functions?

**Answer**

Yes, a View can contain aggregate functions, but like JOINs, this typically makes the View non-updatable.

Q8. Can you update every View?

**Answer**

No. Only simple Views based on a single table without JOINs, aggregates, GROUP BY, or DISTINCT are typically updatable.

Section D : Indexed Views (Concept)

-------------------------------------------------------------------------------------------------------------------------------

What is an Indexed View?

📖 Definition

An Indexed View is a View that has a unique clustered index. Unlike a standard View, SQL Server stores the View's result physically, which can improve performance for certain queries.


-------------------------------------------------------------------------------------------------------------------------------

Important Note

This is an advanced SQL Server feature.

For Data Analyst interviews, you only need to understand the concept.

Standard View vs Indexed View

Standard View	Indexed View

Does not store data	Stores indexed results

Executes the query every time	Reads stored indexed data

Easier to maintain	Requires additional maintenance
