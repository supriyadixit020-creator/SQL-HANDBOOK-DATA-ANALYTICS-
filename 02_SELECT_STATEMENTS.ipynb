1. SELECT Statement

-------------------------------------------------------------------------------------------------------------------------------

The SELECT statement is used to retrieve data from one or more columns of a table in a database. It is the most commonly used SQL statement because it allows users to view and analyze stored data.

The SELECT statement is the most frequently used SQL command because it is used to retrieve data from one or more tables.



-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Use SELECT?

We use the SELECT statement whenever we want to retrieve data from a database table. 

For example:

Show all employees

Show customer names

Show product prices

Show total sales

Show student marks

8. ----------FROM Clause--------------

-------------------------------------------------------------------------------------------------------------------------------


The FROM clause specifies the table from which the data will be retrieved.

Example:

SELECT Name
FROM Employees;

Here,

SELECT → tells SQL what to retrieve.

FROM → tells SQL where to retrieve it from.

9. --------SELECT Execution Flow-------

-------------------------------------------------------------------------------------------------------------------------------

Example:

SELECT Name
FROM Employees;

"SQL first finds the Employees table and then retrieves the Name column."

#Interview Question: Does SQL execute SELECT before FROM?

Answer: No. Logically, SQL processes the FROM clause first and then the SELECT clause

Part 2 : ----------DISTINCT Clause---------------

-------------------------------------------------------------------------------------------------------------------------------

1. Definition 

The DISTINCT keyword is used to return only unique values from a column or a combination of columns. It removes duplicate records from the result set.

-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Use DISTINCT?

Retreive unique values from data when you have duplicate values

-----------------------------------------------------------------------------------------------------------------------------

3. Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT DISTINCT column_name

FROM table_name;

------------------------------------------------------------------------------------------------------------------------------

4. Example 1 – Unique Departments

FROM Employees;

| ---------- |

| HR         |

| IT         |

| Finance    |

11. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is the purpose of the DISTINCT keyword?

Answer

The DISTINCT keyword is used to return only unique values by removing duplicate records from the query result.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Does DISTINCT remove duplicate records from the table?

Answer

No.

It only removes duplicate values from the result set, not from the actual table.

------------------------------------------------------------------------------------------------------------------------------

Q3. Can DISTINCT be used with multiple columns?

Answer

Yes.

It returns unique combinations of the selected columns.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Is DISTINCT a DML command?

Answer

No.

It is a keyword used with the SELECT statement.



-----------------------------------------------------------------------------------------------------------------------------

📚 Quick Revision-----------------------

-------------------------------------------------------------------------------------------------------------------------------

DISTINCT removes duplicate values only in the output.

It does not delete data from the table.

It can be used with one or multiple columns.

With multiple columns, it checks the combination of values.

Use it only when unique results are required.

Part 3 : TOP and LIMIT

-------------------------------------------------------------------------------------------------------------------------------

1. Definition (Interview Ready)

The TOP and LIMIT clauses are used to restrict the number of rows returned by a SQL query. They help retrieve only the required number of records instead of the entire table.

-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Use TOP / LIMIT?

Imagine a table has 10 million records.

You don't always want to display all of them.

Sometimes you only need:

First 5 employees,
Top 10 products,
First 100 customers&
Top 20 highest sales.

Instead of retrieving the entire table, we return only the required rows.

This improves readability and, in many cases, performance.

-------------------------------------------------------------------------------------------------------------------------------

3. SQL Server vs MySQL

-------------------------------------------------------------------------------------------------------------------------------

This is a very common interview question.

| Database             | Clause Used |
| -------------------- | ----------- |
| Microsoft SQL Server | `TOP`       |
| MySQL                | `LIMIT`     |
| PostgreSQL           | `LIMIT`     |
| SQLite               | `LIMIT`     |

==========================

Syntax

FROM table_name;

FROM Employees;

102	Emma	IT	60000

103	David	Finance	55000

-------------------------------------------------------------------------------------------------------------------------------

Example 2

Display the first 2 employee names.

FROM Employees;

| Name |
| ---- |
| John |
| Emma |

-------------------------------------------------------------------------------------------------------------------------------

6. LIMIT Clause (MySQL)

============================

Syntax

SELECT column_name
FROM table_name
LIMIT number;

-------------------------------------------------------------------------------------------------------------------------------

EXAMPLE

SELECT *
FROM Employees
LIMIT 3;

Output is the same as:

SELECT TOP 3 *
FROM Employees;

The only difference is the SQL database being used.

-------------------------------------------------------------------------------------------------------------------------------

7. Important Note

#========================

Without an ORDER BY clause:

SELECT TOP 3 *
FROM Employees;

The database returns the first three rows based on its current order.

📌 This order is not guaranteed.

------------------------------------------------------------------------------------------------------------------------------

8. Using TOP with ORDER BY (Best Practice)

-------------------------------------------------------------------------------------------------------------------------------

Suppose we want the highest-paid employees.

SELECT TOP 3 Name, Salary
FROM Employees
ORDER BY Salary DESC;

OUTPUT

| Name    | Salary |
| ------- | ------ |
| Michael | 65000  |
| Emma    | 60000  |
| David   | 55000  |

Why?

ORDER BY Salary DESC

means:

Highest salary first.

Then

TOP 3

returns only the first three rows from the sorted result.

📌 This is how TOP is commonly used in real projects.

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

-----------------------------------------------------------------------------------------------------------------------------


✅ Always use ORDER BY with TOP or LIMIT when you want the highest, lowest, latest, or oldest records.

Good

FROM Employees;

unless you genuinely just need any first five rows.

13. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is the purpose of TOP?

Answer

The TOP clause is used in SQL Server to limit the number of rows returned by a query.

-------------------------------------------------------------------------------------------------------------------------------

Q2. What is the difference between TOP and LIMIT?

Answer

TOP is used in SQL Server, whereas LIMIT is used in MySQL, PostgreSQL, and SQLite. Both are used to restrict the number of rows returned.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Why should TOP usually be used with ORDER BY?

Answer

ORDER BY ensures that the returned rows are sorted correctly before TOP selects the required number of records.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Does TOP modify the data?

Answer

No.

It only limits the rows displayed in the result set

-------------------------------------------------------------------------------------------------------------------------------

📚 Quick Revision

-------------------------------------------------------------------------------------------------------------------------------

TOP is used in SQL Server.

LIMIT is used in MySQL/PostgreSQL/SQLite.

Both restrict the number of rows returned.

TOP and LIMIT do not modify the data.

Use ORDER BY with TOP or LIMIT to get meaningful results like highest salary or latest orders.

Part 4 : Alias (AS)

-------------------------------------------------------------------------------------------------------------------------------

1. Definition

An alias is a temporary name given to a column or a table in a SQL query to improve readability and make the output easier to understand.

-------------------------------------------------------------------------------------------------------------------------------

3. Syntax

Column Alias

SELECT column_name AS alias_name
FROM table_name;

-------------------------------------------------------------------------------------------------------------------------------

Table Alias

SELECT *
FROM table_name AS alias_name;

-------------------------------------------------------------------------------------------------------------------------------

| 65000           |

Original column

ban gaya.

13. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is an Alias in SQL?

Answer

An alias is a temporary name assigned to a column or table in a SQL query to improve readability and make the output easier to understand.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Does an alias permanently change the column name?

Answer

No.

An alias exists only during query execution.

The original column name remains unchanged.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Is the AS keyword mandatory?

Answer

No.

The AS keyword is optional, but using it is considered a best practice because it improves readability.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Why are table aliases used?

Answer

Table aliases provide a shorter name for a table, making queries easier to read and write, especially when working with joins.

-------------------------------------------------------------------------------------------------------------------------------

📚 Quick Revision

-------------------------------------------------------------------------------------------------------------------------------

Alias = Temporary Name.

Alias can be given to columns and tables.

AS keyword is optional but recommended.

Alias improves readability.

Alias does not change the original table or column name.

Table aliases are especially useful when writing JOIN queries.
