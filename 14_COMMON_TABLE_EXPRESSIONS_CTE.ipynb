1. What is a CTE?


ЁЯУЦ Definition (Interview Ready)

A Common Table Expression (CTE) is a temporary named result set created using the WITH keyword that can be referenced within a single SQL statement. It improves the readability and maintainability of complex queries.

1. What are Multiple CTEs?

ЁЯУЦ Definition (Interview Ready)

Multiple CTEs allow you to define two or more Common Table Expressions in a single SQL statement. Each CTE can be used by the final query, and later CTEs can reference earlier CTEs.

------------------------------------------------------------------------------------------------------------------------------


Syntax

------------------------------------------------------------------------------------------------------------------------------

WITH CTE1 AS

(

    SELECT ...

),


CTE2 AS

(

    SELECT ...

)

SELECT *

FROM CTE2;

------------------------------------------------------------------------------------------------------------------------------

Important Rule

------------------------------------------------------------------------------------------------------------------------------

тЬЕ Separate multiple CTEs with commas (,).

тЭМ Do not write WITH again for the second CTE.

Section E : Employee Hierarchy

-------------------------------------------------------------------------------------------------------------------------------

6. Business Question

Display employee hierarchy.

Query

WITH EmployeeHierarchy AS
(
    -- Anchor Member
    SELECT
        EmployeeID,
        EmployeeName,
        ManagerID,
        1 AS Level
    FROM Employees
    WHERE ManagerID IS NULL

    UNION ALL

    -- Recursive Member
    SELECT
        e.EmployeeID,
        e.EmployeeName,
        e.ManagerID,
        eh.Level + 1
    FROM Employees e
    INNER JOIN EmployeeHierarchy eh
        ON e.ManagerID = eh.EmployeeID
)

SELECT *
FROM EmployeeHierarchy;

10. Common Mistakes

-------------------------------------------------------------------------------------------------------------------------------

тЭМ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Forgetting the stopping condition.

This may cause an infinite recursion error.

-------------------------------------------------------------------------------------------------------------------------------

тЭМ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Using UNION instead of UNION ALL without a specific reason.

UNION ALL avoids unnecessary duplicate elimination and is generally more efficient.

-------------------------------------------------------------------------------------------------------------------------------

тЭМ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Incorrectly joining the Recursive Member.

Always join using the parent-child relationship.

-------------------------------------------------------------------------------------------------------------------------------

тЪб Performance Tip

Recursive CTEs are powerful but can become expensive on very large hierarchies. Ensure there is a proper termination condition and consider indexing parent-child key columns (such as ManagerID) for better performance.
