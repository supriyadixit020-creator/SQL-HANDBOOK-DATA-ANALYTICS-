Section A : Introduction to Temporary Tables

-------------------------------------------------------------------------------------------------------------------------------

1. What is a Temporary Table?
📖 Definition (Interview Ready)

A Temporary Table is a table that is created to store data temporarily during the execution of a SQL session or procedure. It exists only for a limited period and is automatically removed when it is no longer needed.



-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need Temporary Tables?

In real-world projects, we often need to:

Store intermediate results.

Break complex queries into smaller steps.

Improve query readability.

Process large datasets.

Reuse intermediate data multiple times.

Instead of repeatedly executing the same query, we can save the results in a Temporary Table.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Scenario

A retail company wants to generate a monthly sales report.

The original query joins five large tables:

Orders
Customers
Products
Employees
Regions

Instead of joining these tables multiple times, the SQL developer stores the intermediate result in a Temporary Table.

The final report becomes faster and easier to understand.

Section C : Local Temporary Tables

-------------------------------------------------------------------------------------------------------------------------------

2. What is a Local Temporary Table?


📖 Definition (Interview Ready)

A Local Temporary Table is a temporary table that is visible only within the current database session or connection. It is automatically dropped when the session ends.



-------------------------------------------------------------------------------------------------------------------------------

Syntax

CREATE TABLE #TableName
(
Column1 DataType,
Column2 DataType
);

Business Requirement

The Sales Manager wants to analyze only the current month's sales.

Instead of creating a permanent table, create a Temporary Table.

Query

CREATE TABLE #MonthlySales
(
SaleID INT,
ProductName VARCHAR(100),
SalesAmount DECIMAL(10,2)
);
Insert Data
INSERT INTO #MonthlySales
VALUES
(1,'Laptop',65000),
(2,'Keyboard',3500),
(3,'Mouse',1500);
Display Data
SELECT *
FROM #MonthlySales;

Output

SaleID	ProductName	SalesAmount

1	Laptop	65000.00

2	Keyboard	3500.00

3	Mouse	1500.00

Step-by-Step Explanation
Step 1

Create the Temporary Table.

↓

SQL creates the table in the tempdb database.

Step 2

Insert temporary data.

↓

The data is stored only for the current session.

Step 3

Run queries.

↓

SQL treats it like a normal table.

Step 4

Close the session.

↓

SQL automatically removes the Temporary Table.

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a Temporary Table?

**Answer**

A Temporary Table is a table that is created and stored temporarily in the database, used to hold intermediate results during a session or process, and is automatically dropped afterward.

Q2. What is a Local Temporary Table?

**Answer**

A Local Temporary Table (prefixed with #) is visible only within the session that created it and is automatically dropped when that session ends.

Q3. Where are Temporary Tables stored?

**Answer**

Temporary Tables are stored in the database's temporary storage area (e.g., tempdb in SQL Server), separate from the regular user database.

Q4. When is a Local Temporary Table deleted?

**Answer**

A Local Temporary Table is automatically deleted when the session that created it closes, or when it is explicitly dropped using DROP TABLE.

Q5. Why are Temporary Tables commonly used in ETL processes?

**Answer**

They are used in ETL processes to hold intermediate, staged data during multi-step transformations before the final data is loaded into the target table.

Section A : Global Temporary Tables

-------------------------------------------------------------------------------------------------------------------------------

1. What is a Global Temporary Table?

📖 Definition (Interview Ready)

A Global Temporary Table is a temporary table that is accessible to all database sessions. It remains available until the session that created it ends and no other active session is using it.


-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need Global Temporary Tables?

Sometimes multiple users or SQL sessions need to work with the same temporary data.

Instead of creating separate Local Temporary Tables, a Global Temporary Table allows everyone to access the same temporary dataset.

Syntax

CREATE TABLE ##TableName

(

Column1 DataType,

Column2 DataType

);


💼 Business Requirement

The Finance team generates a temporary monthly report.

The HR team also needs to access the same report during the day.

A Global Temporary Table allows both teams to use the same temporary data.

Query
CREATE TABLE ##MonthlyReport

Salary DECIMAL(10,2)

);

Insert Data

Section B : Table Variables

2. What is a Table Variable?


📖 Definition (Interview Ready)

A Table Variable is a special type of variable that stores rows and columns temporarily within the current batch, stored procedure, or function.

-------------------------------------------------------------------------------------------------------------------------------

Why Do We Need Table Variables?

Sometimes we only need a small amount of temporary data inside a single query or stored procedure.

Instead of creating a Temporary Table, we can use a Table Variable.

Syntax

(

EmployeeName VARCHAR(100),

Salary DECIMAL(10,2)

);

🎤 Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is a Global Temporary Table?

**Answer**

A Global Temporary Table (prefixed with ##) is visible to all sessions and is dropped only when the session that created it ends and no other session is referencing it.

Q2. What is a Table Variable?

**Answer**

A Table Variable (declared with DECLARE @TableName TABLE) is a variable that stores data temporarily in memory, scoped to the batch or procedure in which it is declared.

Q3. What is the difference between # and ##?

**Answer**

A single # denotes a Local Temporary Table visible only to the current session, while ## denotes a Global Temporary Table visible to all sessions.

Q4. What is the difference between a Temporary Table and a Table Variable?

**Answer**

Temporary Tables support indexes, constraints, and are logged more extensively, making them better for large datasets, while Table Variables are lighter-weight, scoped to the batch, and better suited for small datasets.

Q5. Where are Temporary Tables stored?

**Answer**

Temporary Tables are stored in the database's temporary storage area (e.g., tempdb in SQL Server), separate from the regular user database.

Q6. Which option is better for large datasets?

**Answer**

Temporary Tables are generally better for large datasets since they support indexing and statistics, which improves query performance.

Q7. Which option is better for small datasets?

**Answer**

Table Variables are generally better for small datasets due to their lower overhead.

Business Requirement

A company generates a monthly employee performance report.

The report requires multiple calculations before displaying the final result.

Instead of repeating calculations, use a Temporary Table inside the Stored Procedure.

Example

(

Section F : SQL Server Interview Tips

-------------------------------------------------------------------------------------------------------------------------------

⭐ Tip 1

If the interviewer asks:

"When would you use a Temporary Table instead of a CTE?"

Answer:

Use a Temporary Table when the intermediate result needs to be reused multiple times or when working with large datasets. Use a CTE for a one-time calculation within a single query.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Tip 2

If asked:

"When would you use a Table Variable?"

Answer:

Use a Table Variable for relatively small datasets inside stored procedures or functions where only temporary storage is required.

-------------------------------------------------------------------------------------------------------------------------------

⭐ Tip 3

If asked:

"Can Temporary Tables be indexed?"

Answer:

Yes. Temporary Tables can have indexes, primary keys, and constraints, just like regular tables.
