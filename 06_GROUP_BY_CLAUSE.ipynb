Definition (Interview Ready)

The GROUP BY clause is used to group rows that have the same values in one or more columns. It is commonly used with Aggregate Functions to calculate summary values for each group.
GROUP BY groups rows having the same values into summary rows.

-------------------------------------------------------------------------------------------------------------------------------

Why Do We Use GROUP BY?

Aggregate Functions calculate a summary for the entire table.

| Company | GROUP BY Use        |

| ------- | ------------------- |

| Amazon  | Sales by Category   |

| Netflix | Movies by Genre     |

| Swiggy  | Orders by City      |

| Bank    | Customers by Branch |

-------------------------------------------------------------------------------------------------------------------------------

Execution flow :

-------------------------------------------------------------------------------------------------------------------------------

FROM

↓

WHERE

↓

GROUP BY

↓

Aggregate Function

↓

SELECT

↓

ORDER BY

-------------------------------------------------------------------------------------------------------------------------------

Example:

SELECT AVG(Salary)
FROM Employees;

Output

AVG(Salary)

53833.33

This gives the average salary of the entire company.

But what if the HR Manager asks:

What is the average salary in each department?

How many employees work in each department?

What is the highest salary in each department?

A single aggregate function cannot answer these questions.

This is where GROUP BY is used.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

Imagine a company with 20 departments.

The HR Director wants to compare the average salary of every department to identify salary differences.

Instead of creating 20 separate queries, one GROUP BY query can generate the complete report.

4. GROUP BY with COUNT()

-------------------------------------------------------------------------------------------------------------------------------

📖 Business Question

-------------------------------------------------------------------------------------------------------------------------------

How many employees work in each department?

Query

SELECT Department,
       COUNT(*) AS Total_Employees
FROM Employees
GROUP BY Department;

Output

Department	Total_Employees

Finance	1

HR	2

IT	2

Marketing	1

Explanation

SQL groups employees by Department.

Then it counts the number of employees in each group.

-------------------------------------------------------------------------------------------------------------------------------

   ↓

   ↓

9. Important Rules

-------------------------------------------------------------------------------------------------------------------------------

Rule 1

Every column in the SELECT list must either:

Be included in the GROUP BY clause, or

Be inside an Aggregate Function.

-------------------------------------------------------------------------------------------------------------------------------

Correct Example

SELECT Department,
       COUNT(*)
FROM Employees
GROUP BY Department;

Wrong Example

SELECT Name,
       Department,
       COUNT(*)
FROM Employees
GROUP BY Department;

❌ This produces an error because Name is neither grouped nor aggregated.

12. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. What is the purpose of the GROUP BY clause?

Answer

The GROUP BY clause groups rows with the same values and is commonly used with Aggregate Functions to calculate summary values for each group.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Can GROUP BY be used without Aggregate Functions?

Answer

Yes. It groups duplicate values into unique groups, but it is most commonly used with Aggregate Functions.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Why is GROUP BY used with Aggregate Functions?

Answer

Aggregate Functions calculate a single summary value, whereas GROUP BY allows them to calculate separate summaries for each group.

-------------------------------------------------------------------------------------------------------------------------------

Q4. Does GROUP BY sort the data?

Answer

No. Use ORDER BY if you need the result sorted.

-------------------------------------------------------------------------------------------------------------------------------

-------------------------------------------------------------------------------------------------------------------------------

✅ It is commonly used with Aggregate Functions.

✅ One result is returned for each group.

✅ Every selected column must be grouped or aggregated.

✅ GROUP BY does not sort the result.

1. GROUP BY with WHERE

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The WHERE clause filters rows before the GROUP BY clause creates groups.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Find the average salary of only IT employees.

Query

SELECT Department,
       AVG(Salary) AS Average_Salary
FROM Employees
WHERE Department = 'IT'
GROUP BY Department;

Output

Department	Average_Salary

IT	62500

Explanation

Step 1 → Filter IT employees.

Step 2 → Create the IT group.

Step 3 → Calculate the average salary.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------

Managers often request reports for a specific department instead of the whole company.

-------------------------------------------------------------------------------------------------------------------------------

Example:

Average salary of IT

Total sales of Delhi

Employees in HR

2. GROUP BY with ORDER BY

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

The ORDER BY clause sorts the grouped result after the groups and aggregate values have been created.

-------------------------------------------------------------------------------------------------------------------------------

GROUP BY vs ORDER BY

-------------------------------------------------------------------------------------------------------------------------------

Comparison Table.

GROUP BY | ORDER BY

Groups data	 | Sorts data

Aggregate use	| No aggregate required

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Display departments by highest average salary.

-------------------------------------------------------------------------------------------------------------------------------

Query

SELECT Department,
       AVG(Salary) AS Average_Salary
FROM Employees
GROUP BY Department
ORDER BY Average_Salary DESC;

Output

Department	Average_Salary

IT	62500

Finance	55000

Marketing	50000

HR	46500

-------------------------------------------------------------------------------------------------------------------------------

Explanation

SQL groups employees by department.

Calculates the average salary.

Sorts the departments from highest to lowest average salary.

-------------------------------------------------------------------------------------------------------------------------------

💼 Business Insight

-------------------------------------------------------------------------------------------------------------------------------


This type of report helps HR identify departments with higher salary costs.

3. GROUP BY with Multiple Columns

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition (Interview Ready)

The GROUP BY clause can group data using two or more columns. Each unique combination forms a separate group.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------


Count employees by Department and City.

Query

SELECT Department,
       City,
       COUNT(*) AS Total_Employees
FROM Employees
GROUP BY Department, City;

Output

Department	City	Total_Employees

Finance	Delhi	1

HR	Delhi	1

HR	Lucknow	1

IT	Mumbai	2

Marketing	Pune	1

Explanation

Each unique combination becomes a separate group.

Examples:

HR + Delhi

HR + Lucknow

IT + Mumbai

-------------------------------------------------------------------------------------------------------------------------------

🎯 Interview Tip

-------------------------------------------------------------------------------------------------------------------------------


SQL groups data from left to right.

GROUP BY Department, City;

First by Department, then by City.

4. COUNT(DISTINCT column)

-------------------------------------------------------------------------------------------------------------------------------

📖 Definition

COUNT(DISTINCT column_name) counts only unique non-NULL values in a column.

-------------------------------------------------------------------------------------------------------------------------------

Business Question

-------------------------------------------------------------------------------------------------------------------------------

How many different departments are there?

Query

SELECT COUNT(DISTINCT Department) AS Total_Departments
FROM Employees;

Output

Total_Departments

4

Explanation

Departments are:

HR

IT

Finance

Marketing

Although HR and IT appear multiple times, they are counted only once.

Useful for finding:

8. Interview Questions

-------------------------------------------------------------------------------------------------------------------------------

Q1. Which clause executes first: WHERE or GROUP BY?

Answer

The WHERE clause executes first. It filters rows before the GROUP BY clause creates groups.

-------------------------------------------------------------------------------------------------------------------------------

Q2. Can ORDER BY be used with GROUP BY?

Answer

Yes. After groups and aggregate values are created, ORDER BY can sort the grouped result.

-------------------------------------------------------------------------------------------------------------------------------

Q3. Can GROUP BY use multiple columns?

Answer

Yes. SQL creates one group for each unique combination of the specified columns.

-------------------------------------------------------------------------------------------------------------------------------

Q4. What is the purpose of COUNT(DISTINCT column_name)?

Answer

It counts only unique non-NULL values in the specified column.
