1. Definition:

The WHERE clause is used to filter records in a SQL query. It returns only those rows that satisfy a specified condition.
The WHERE clause is used to filter records based on a specified condition


-------------------------------------------------------------------------------------------------------------------------------

2. Why Do We Use the WHERE Clause?

When working with databases, we usually do not need all the records.

Instead, we need only the records that match a specific condition.

| Company  | WHERE Use           |

| -------- | ------------------- |

| Amazon   | Price > ₹1000 products |

| Swiggy   | Veg restaurants        |

| Bank     | Balance > ₹50,000      |

| Hospital | Age > 60 patients      |


-------------------------------------------------------------------------------------------------------------------------------

Example-------

Instead of displaying all employees,

we may want to display:

Employees from the IT department

Employees earning more than ₹50,000

Customers from Delhi

Orders placed in 2025

NOTE:--- The WHERE clause helps us retrieve only the required data.

-------------------------------------------------------------------------------------------------------------------------------

3. Real-World Example

-------------------------------------------------------------------------------------------------------------------------------

Imagine a company has 50,000 employees.

The HR manager asks:

"Show only employees who work in the IT department."

Instead of viewing all employees, SQL filters only the required records.

This is where the WHERE clause is used.

-------------------------------------------------------------------------------------------------------------------------------

4. Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT column_name
FROM table_name
WHERE condition;

-------------------------------------------------------------------------------------------------------------------------------

Syntax Breakdown

===================

| Keyword   | Purpose                                |
| --------- | -------------------------------------- |
| SELECT    | Specifies the columns to retrieve.     |
| FROM      | Specifies the table name.              |
| WHERE     | Filters the records.                   |
| Condition | Defines which rows should be returned. |

Example 1 – Display Employees from the IT Department

-------------------------------------------------------------------------------------------------------------------------------

Query

SELECT *
FROM Employees
WHERE Department = 'IT';

Output

EmployeeID	Name	Department	Salary	City

102	Emma	IT	60000	Mumbai

105	Michael	IT	65000	Mumbai

Explanation

SQL first reads the Employees table.

It checks every row.

It returns only the rows where Department = 'IT'.

-------------------------------------------------------------------------------------------------------------------------------

How SQL Processes This Query

-------------------------------------------------------------------------------------------------------------------------------

-------------------Consider the query:

SELECT Name
FROM Employees
WHERE Department = 'IT';

-------------------------------------------------------------------------------------------------------------------------------

Logical Execution Order
FROM
   ↓
WHERE
   ↓
SELECT

------------------------------------------------------------------------------------------------------------------------------
Explanation

SQL identifies the Employees table.

It filters rows where Department = 'IT'.

Finally, it returns only the Name column

------------------------------------------------------------------------------------------------------------------------------

📌 Interview Tip: SQL logically processes FROM first, then WHERE, and finally SELECT.

------------------------------------------------------------------------------------------------------------------------------

Best Practices

------------------------------------------------------------------------------------------------------------------------------

✅ Write SQL keywords in uppercase.

SELECT Name
FROM Employees
WHERE Department = 'IT';

------------------------------------------------------------------------------------------------------------------------------

✅ Write meaningful conditions.

Good:

WHERE Department = 'IT'

Avoid vague or incorrect conditions.

------------------------------------------------------------------------------------------------------------------------------

✅ Use single quotes for text values.

Correct:

WHERE City = 'Delhi';

------------------------------------------------------------------------------------------------------------------------------


Incorrect:

WHERE City = Delhi;

------------------------------------------------------------------------------------------------------------------------------


📌 Rule: Strings (text values) should be enclosed in single quotes.

Interview Questions

------------------------------------------------------------------------------------------------------------------------------

Q1. What is the purpose of the WHERE clause?

Answer

The WHERE clause is used to filter records and return only the rows that satisfy a specified condition.

------------------------------------------------------------------------------------------------------------------------------

Q2. Can the WHERE clause be used without SELECT?

Answer

No.

The WHERE clause is used along with statements like SELECT, UPDATE, or DELETE to specify conditions.

------------------------------------------------------------------------------------------------------------------------------

Q3. Does the WHERE clause modify data?

Answer

No.

It only filters the result set. It does not change the data stored in the table.

------------------------------------------------------------------------------------------------------------------------------

Q4. Why are single quotes used in the WHERE clause?

Answer

Single quotes are used to represent string (text) values, while numeric values do not require quotes

 ↓

 ↓

Chapter 3 : WHERE Clause (Part 2)

Comparison Operators

------------------------------------------------------------------------------------------------------------------------------

Definition 

Comparison operators are used in the WHERE clause to compare two values. They help filter records based on specific conditions

------------------------------------------------------------------------------------------------------------------------------

Why Do We Use Comparison Operators?

Suppose a manager asks:

Show employees earning more than ₹50,000.

Show products cheaper than ₹1,000.

Show customers whose age is 25.

Show employees whose salary is not ₹60,000.

These questions cannot be answered using only the WHERE clause.

We need comparison operators.

------------------------------------------------------------------------------------------------------------------------------

Comparison Operators Table

------------------------------------------------------------------------------------------------------------------------------

Operator	Meaning	Example

=	Equal to	Salary = 50000

>	Greater than	Salary > 50000

<	Less than	Salary < 50000

>=	Greater than or equal to	Salary >= 50000

<=	Less than or equal to	Salary <= 50000

!= or <>	Not equal to	Department != 'HR'

------------------------------------------------------------------------------------------------------------------------------

📌 Interview Tip:\

------------------------------------------------------------------------------------------------------------------------------

Both != and <> mean Not Equal To. <> is the ANSI SQL standard, while SQL Server supports both.

------------------------------------------------------------------------------------------------------------------------------

Equal To (=)

Definition

The = operator returns rows where the column value exactly matches the specified value.

------------------------------------------------------------------------------------------------------------------------------

Example 1

Query

SELECT *
FROM Employees
WHERE Department = 'IT';

Output

EmployeeID	Name	Department	Salary

102	Emma	IT	60000

105	Michael	IT	65000

------------------------------------------------------------------------------------------------------------------------------

Explanation

------------------------------------------------------------------------------------------------------------------------------

Only employees whose Department is IT are returned.

------------------------------------------------------------------------------------------------------------------------------

Greater Than (>)

------------------------------------------------------------------------------------------------------------------------------

Definition

The > operator returns rows where the column value is greater than the specified value.

Example

SELECT Name, Salary
FROM Employees
WHERE Salary > 50000;

Output

Name	Salary

Emma	60000

David	55000

Michael	65000

------------------------------------------------------------------------------------------------------------------------------

Explanation

------------------------------------------------------------------------------------------------------------------------------

Employees earning more than ₹50,000 are displayed.

------------------------------------------------------------------------------------------------------------------------------

6. Less Than (<)

------------------------------------------------------------------------------------------------------------------------------

Definition

The < operator returns rows where the column value is less than the specified value.

Example

SELECT Name, Salary
FROM Employees
WHERE Salary < 50000;

Output

Name	Salary

John	45000

Olivia	48000

------------------------------------------------------------------------------------------------------------------------------

7. Greater Than or Equal To (>=)

------------------------------------------------------------------------------------------------------------------------------

Definition

The >= operator returns rows where the column value is greater than or equal to the specified value.

Example

SELECT Name, Salary
FROM Employees
WHERE Salary >= 50000;

Output

Name	Salary

Emma	60000

David	55000

Sophia	50000

Michael	65000

------------------------------------------------------------------------------------------------------------------------------

📌 Notice that Sophia is included because her salary is exactly 50000.

------------------------------------------------------------------------------------------------------------------------------

8. Less Than or Equal To (<=)

------------------------------------------------------------------------------------------------------------------------------

Definition

The <= operator returns rows where the column value is less than or equal to the specified value.

Example

SELECT Name, Salary
FROM Employees
WHERE Salary <= 50000;

Output

Name	Salary

John	45000

Sophia	50000

Olivia	48000

------------------------------------------------------------------------------------------------------------------------------

9. Not Equal To (!= / <>)

------------------------------------------------------------------------------------------------------------------------------

Definition

The != or <> operator returns rows where the column value is not equal to the specified value.

Example

SELECT Name, Department
FROM Employees
WHERE Department != 'HR';

Output

Name	Department

Emma	IT

David	Finance

Sophia	Marketing

Michael	IT

------------------------------------------------------------------------------------------------------------------------------

The same query can also be written as:

------------------------------------------------------------------------------------------------------------------------------

SELECT Name, Department
FROM Employees
WHERE Department <> 'HR';

Both queries return the same result.

13. Interview Questions

------------------------------------------------------------------------------------------------------------------------------

Q1. What are comparison operators in SQL?

Answer

Comparison operators are used in the WHERE clause to compare values and filter records based on specified conditions.

------------------------------------------------------------------------------------------------------------------------------

Q2. What is the difference between > and >=?

Answer

 returns values strictly greater than the specified value, whereas >= returns values greater than or equal to the specified value.

------------------------------------------------------------------------------------------------------------------------------

Chapter 3 - WHERE Clause

Part 3 - Logical Operators (AND, OR, NOT)

------------------------------------------------------------------------------------------------------------------------------

Definition 

Logical operators are used to combine one or more conditions in the WHERE clause. They help filter data based on multiple criteria.

------------------------------------------------------------------------------------------------------------------------------

Why Do We Use Logical Operators?

Sometimes, a single condition is not enough.

------------------------------------------------------------------------------------------------------------------------------

For example:

Show employees from the IT department and earning more than ₹50,000.

Show employees from HR or Finance.

Show employees not working in the Marketing department.

To solve such business problems, we use Logical Operators.

------------------------------------------------------------------------------------------------------------------------------

Types of Logical Operators

------------------------------------------------------------------------------------------------------------------------------

Operator	Purpose

AND	Returns rows only if all conditions are true.

OR	Returns rows if at least one condition is true.

NOT	Reverses a condition and returns rows that do not satisfy it.

------------------------------------------------------------------------------------------------------------------------------

AND Operator

------------------------------------------------------------------------------------------------------------------------------

Definition

The AND operator returns rows only when all specified conditions are true.

------------------------------------------------------------------------------------------------------------------------------

Think Like This

Condition 1 ✅

AND

Condition 2 ✅

=

Result ✅

If any one condition is false, the row is not returned.

------------------------------------------------------------------------------------------------------------------------------

Syntax

SELECT column_name

FROM table_name

WHERE condition1

AND condition2;

------------------------------------------------------------------------------------------------------------------------------

Example 1

Business Question

Show employees who work in the IT department and earn more than ₹60,000.

Query

SELECT *

FROM 

WHERE Department = 'IT'

AND Salary > 60000;

Output

EmployeeID	Name	Department	Salary

105	Michael	IT	65000

Explanation

Emma works in IT but earns exactly ₹60,000, not more than ₹60,000.

Michael satisfies both conditions.

------------------------------------------------------------------------------------------------------------------------------

OR Operator

------------------------------------------------------------------------------------------------------------------------------

Definition

The OR operator returns rows when at least one of the specified conditions is true.

------------------------------------------------------------------------------------------------------------------------------

Think Like This

Condition 1 ✅

OR

Condition 2 ❌

=

Result ✅

Only one condition needs to be true.

------------------------------------------------------------------------------------------------------------------------------

Syntax

SELECT column_name

FROM table_name

WHERE condition1

OR condition2;

------------------------------------------------------------------------------------------------------------------------------

Example 1

Business Question

Show employees from the HR or Finance department.

Query

SELECT Name, Department

FROM Employees

WHERE Department = 'HR'

OR Department = 'Finance';

Output

Name	Department

John	HR

David	Finance

Olivia	HR

------------------------------------------------------------------------------------------------------------------------------

NOT Operator

------------------------------------------------------------------------------------------------------------------------------

Definition

The NOT operator reverses a condition and returns rows that do not satisfy the specified condition.

------------------------------------------------------------------------------------------------------------------------------

Syntax

SELECT column_name

FROM table_name

WHERE NOT condition;

------------------------------------------------------------------------------------------------------------------------------

Example 1

Business Question

Show employees who are not in the HR department.

Query

SELECT Name, Department

FROM Employees

WHERE NOT Department = 'HR';

Output

Name	Department

Emma	IT

David	Finance

Sophia	Marketing

Michael	IT

------------------------------------------------------------------------------------------------------------------------------

Combining AND and OR

------------------------------------------------------------------------------------------------------------------------------

This is a very common interview topic.

------------------------------------------------------------------------------------------------------------------------------

Example

Show employees who:

Work in the IT department OR

Earn more than ₹60,000.

SELECT Name, Department, Salary

FROM Employees

WHERE Department = 'IT'

OR Salary > 60000;

Output

Name	Department	Salary

Emma	IT	60000

Michael	IT	65000

------------------------------------------------------------------------------------------------------------------------------

8. Operator Precedence

------------------------------------------------------------------------------------------------------------------------------

SQL evaluates conditions in this order:

NOT
AND
OR

------------------------------------------------------------------------------------------------------------------------------

📌 Interview Tip: If your query contains both AND and OR, use parentheses to make the logic clear.

------------------------------------------------------------------------------------------------------------------------------

Example

SELECT *

FROM Employees

WHERE (Department = 'IT' OR Department = 'HR')

AND Salary > 50000;

------------------------------------------------------------------------------------------------------------------------------

Explanation

------------------------------------------------------------------------------------------------------------------------------

SQL first evaluates:

Department = 'IT'

OR

Department = 'HR'


Then it checks:

Salary > 50000

------------------------------------------------------------------------------------------------------------------------------

9. Best Practices

------------------------------------------------------------------------------------------------------------------------------

✅ Use parentheses when combining multiple logical operators.

Good

SELECT *

FROM Employees

WHERE (Department = 'IT' OR Department = 'HR')

AND Salary > 50000;

------------------------------------------------------------------------------------------------------------------------------

✅ Format conditions on separate lines for better readability.

SELECT *

FROM Employees

WHERE Department = 'IT'

      AND Salary > 50000;

------------------------------------------------------------------------------------------------------------------------------

10. Common Mistakes

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

Writing

WHERE Department = 'IT'

OR 'HR';

------------------------------------------------------------------------------------------------------------------------------

✅ Correct

WHERE Department = 'IT'

OR Department = 'HR';

------------------------------------------------------------------------------------------------------------------------------

Every condition must include the column name.

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

Ignoring operator precedence.

------------------------------------------------------------------------------------------------------------------------------

Wrong

WHERE Department='IT'

OR Department='HR'

AND Salary>50000;

------------------------------------------------------------------------------------------------------------------------------

This may not return the expected result.

------------------------------------------------------------------------------------------------------------------------------

Better

WHERE (Department='IT'

OR Department='HR')

AND Salary>50000;

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

Using AND when OR is required.

------------------------------------------------------------------------------------------------------------------------------

Example

WHERE City='Delhi'

AND City='Mumbai';

This will return no rows, because one employee cannot belong to both cities at the same time.

------------------------------------------------------------------------------------------------------------------------------

11. Interview Questions

------------------------------------------------------------------------------------------------------------------------------
Q1. What is the difference between AND and OR?

Answer

The AND operator returns rows only when all conditions are true, whereas the OR operator returns rows if at least one condition is true.

------------------------------------------------------------------------------------------------------------------------------

Q2. What does the NOT operator do?

Answer

The NOT operator reverses a condition and returns rows that do not satisfy the specified condition.

------------------------------------------------------------------------------------------------------------------------------

Q3. What is the operator precedence in SQL?

Answer

SQL evaluates conditions in the following order:

NOT
AND
OR

------------------------------------------------------------------------------------------------------------------------------

Q4. Why should parentheses be used?

Answer

Parentheses improve readability and ensure SQL evaluates conditions in the intended order.

------------------------------------------------------------------------------------------------------------------------------

Key Takeaways

------------------------------------------------------------------------------------------------------------------------------

✅ AND requires all conditions to be true.

✅ OR requires at least one condition to be true.

✅ NOT reverses a condition.

✅ SQL evaluates NOT → AND → OR.

✅ Use parentheses when combining multiple logical operators.

Special Filtering Operators

1. Introduction

Sometimes writing multiple conditions makes SQL queries longer and harder to read.

------------------------------------------------------------------------------------------------------------------------------


For example,

Instead of writing:

WHERE Department='HR'
OR Department='IT'
OR Department='Finance'

------------------------------------------------------------------------------------------------------------------------------

We can simply write:

------------------------------------------------------------------------------------------------------------------------------


WHERE Department IN ('HR','IT','Finance');

This is much cleaner and easier to maintain.

PART A : BETWEEN

------------------------------------------------------------------------------------------------------------------------------

2. Definition (Interview Ready)

The BETWEEN operator is used to filter records within a specified range. It includes both the starting and ending values.

------------------------------------------------------------------------------------------------------------------------------

Syntax

-------------------------------------------------------------------------------------------------------------------------------

SELECT column_name
FROM table_name
WHERE column_name BETWEEN value1 AND value2;

------------------------------------------------------------------------------------------------------------------------------

Example 1

Business Question

Show employees whose salary is between ₹50,000 and ₹65,000.

Query

SELECT Name, Salary
FROM Employees
WHERE Salary BETWEEN 50000 AND 65000;

Output

Name	Salary

Emma	60000

David	55000

Sophia	50000

Michael	65000

------------------------------------------------------------------------------------------------------------------------------

Explanation

------------------------------------------------------------------------------------------------------------------------------


Both 50000 and 65000 are included because BETWEEN is inclusive.

------------------------------------------------------------------------------------------------------------------------------

Best Practice

------------------------------------------------------------------------------------------------------------------------------

Instead of

WHERE Salary>=50000
AND Salary<=65000;

------------------------------------------------------------------------------------------------------------------------------

Write

------------------------------------------------------------------------------------------------------------------------------

WHERE Salary BETWEEN 50000 AND 65000;

Much cleaner.

------------------------------------------------------------------------------------------------------------------------------


PART B : NOT BETWEEN

------------------------------------------------------------------------------------------------------------------------------

Definition

The NOT BETWEEN operator returns values outside the specified range.

------------------------------------------------------------------------------------------------------------------------------


Example

SELECT Name, Salary
FROM Employees
WHERE Salary NOT BETWEEN 50000 AND 65000;

------------------------------------------------------------------------------------------------------------------------------

Output

Name	Salary

John	45000

Olivia	48000

------------------------------------------------------------------------------------------------------------------------------

PART C : IN

------------------------------------------------------------------------------------------------------------------------------

Definition 

The IN operator is used to compare a column against multiple values in a single condition.

------------------------------------------------------------------------------------------------------------------------------

Syntax

SELECT column_name
FROM table_name
WHERE column_name IN(value1,value2,value3);

------------------------------------------------------------------------------------------------------------------------------
Example

Business Question

Show employees from the HR or IT department.

Query

SELECT Name, Department
FROM Employees
WHERE Department IN ('HR','IT');

Output

Name	Department

John	HR

Emma	IT

Michael	IT

Olivia	HR

Why IN is Better

------------------------------------------------------------------------------------------------------------------------------
Instead of

------------------------------------------------------------------------------------------------------------------------------

WHERE Department='HR'
OR Department='IT';

------------------------------------------------------------------------------------------------------------------------------

Use

WHERE Department IN ('HR','IT');

Cleaner and easier to read.

------------------------------------------------------------------------------------------------------------------------------

PART D : NOT IN

------------------------------------------------------------------------------------------------------------------------------

Definition

The NOT IN operator returns records whose values are not present in the specified list.

------------------------------------------------------------------------------------------------------------------------------

Example

SELECT Name, Department
FROM Employees
WHERE Department NOT IN ('HR','IT');

Output

Name	Department

David	Finance

Sophia	Marketing

------------------------------------------------------------------------------------------------------------------------------

PART E : LIKE

------------------------------------------------------------------------------------------------------------------------------

Definition 

The LIKE operator is used to search for a specific pattern in a text column.

------------------------------------------------------------------------------------------------------------------------------

Wildcards

------------------------------------------------------------------------------------------------------------------------------

Wildcard	Meaning

%	Zero or more characters

_	Exactly one character

------------------------------------------------------------------------------------------------------------------------------

Example 1

Show employees whose name starts with M.

SELECT *
FROM Employees
WHERE Name LIKE 'M%';

Output
Name
Michael

------------------------------------------------------------------------------------------------------------------------------

Example 2

Show employees whose name ends with a.

SELECT *
FROM Employees
WHERE Name LIKE '%a';

Output
Name
Emma
Sophia

------------------------------------------------------------------------------------------------------------------------------

Example 3

Show employees whose name contains av.

SELECT *
FROM Employees
WHERE Name LIKE '%av%';

Output
Name
David

------------------------------------------------------------------------------------------------------------------------------

Example 4

Names having exactly four letters.

SELECT *
FROM Employees
WHERE Name LIKE '____';

Output

Name

John

Emma

------------------------------------------------------------------------------------------------------------------------------

Each underscore (_) represents exactly one character.

------------------------------------------------------------------------------------------------------------------------------

PART F : NOT LIKE

------------------------------------------------------------------------------------------------------------------------------

Definition

The NOT LIKE operator returns records that do not match a specified pattern.

------------------------------------------------------------------------------------------------------------------------------

Example


SELECT *
FROM Employees
WHERE Name NOT LIKE 'M%';

Output

All employees except Michael.

Real-World Business Examples

------------------------------------------------------------------------------------------------------------------------------

Example 1

Find products priced between ₹1000 and ₹5000.

SELECT ProductName, Price
FROM Products
WHERE Price BETWEEN 1000 AND 5000;

------------------------------------------------------------------------------------------------------------------------------

Example 2

Find customers from Delhi, Mumbai, or Pune.

SELECT CustomerName, City
FROM Customers
WHERE City IN ('Delhi','Mumbai','Pune');

------------------------------------------------------------------------------------------------------------------------------

Example 3

Find products whose names start with A.

SELECT ProductName
FROM Products
WHERE ProductName LIKE 'A%';

Best Practices

------------------------------------------------------------------------------------------------------------------------------

✅ Use BETWEEN instead of two comparison operators whenever possible.

✅ Use IN instead of multiple OR conditions.

✅ Use LIKE only for text columns.

✅ Always use meaningful formatting.

------------------------------------------------------------------------------------------------------------------------------

Good

SELECT Name
FROM Employees
WHERE Department IN ('HR','IT');

Common Mistakes

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 1

-------------------------------------------------------------------------------------------------------------------------------

Wrong

WHERE Salary BETWEEN 65000 AND 50000;

Always write the lower value first.

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 2

-------------------------------------------------------------------------------------------------------------------------------

Wrong

WHERE Department IN ('HR');

If checking only one value, simply use:

WHERE Department='HR';

------------------------------------------------------------------------------------------------------------------------------

❌ Mistake 3

-------------------------------------------------------------------------------------------------------------------------------

Confusing % and _

'M%'

Starts with M.

'%M'

Ends with M.

'%M%'

Contains M.

'_M%'

Interview Questions

------------------------------------------------------------------------------------------------------------------------------

Q1. What is the difference between IN and BETWEEN?

Answer

IN is used to compare a value with multiple discrete values, whereas BETWEEN is used to filter values within a continuous range.

------------------------------------------------------------------------------------------------------------------------------

Q2. Is BETWEEN inclusive?

Answer

Yes.

It includes both the starting and ending values.

------------------------------------------------------------------------------------------------------------------------------

Q3. What are the wildcards used with LIKE?

Answer

% → Zero or more characters
_ → Exactly one character

------------------------------------------------------------------------------------------------------------------------------

Q4. Why is IN preferred over multiple OR conditions?

Answer

IN makes queries shorter, cleaner, easier to read, and easier to maintain.

------------------------------------------------------------------------------------------------------------------------------

📌 Key Takeaways

------------------------------------------------------------------------------------------------------------------------------

✅ BETWEEN filters a range (inclusive).

✅ NOT BETWEEN filters values outside a range.

✅ IN compares multiple values in one condition.

✅ NOT IN excludes specific values.

✅ LIKE searches text patterns using % and _.

✅ NOT LIKE excludes matching patterns.

✅ IN is cleaner than multiple OR conditions.

Every condition must reference a column.

------------------------------------------------------------------------------------------------------------------------------


❌ Mistake 4

WHERE Salary BETWEEN 70000 AND 50000;

Always write the lower value first.

------------------------------------------------------------------------------------------------------------------------------


❌ Mistake 5

Ignoring parentheses.

Bad

WHERE Department='IT'
OR Department='HR'
AND Salary>50000;

Better

WHERE (Department='IT'
OR Department='HR')
AND Salary>50000;

🎤 Interview Questions

------------------------------------------------------------------------------------------------------------------------------

Q1. What is the purpose of the WHERE clause?

Answer

The WHERE clause is used to filter records and return only the rows that satisfy a specified condition.

------------------------------------------------------------------------------------------------------------------------------


Q2. What is the difference between WHERE and HAVING?

Answer

The WHERE clause filters rows before grouping, whereas the HAVING clause filters groups after the GROUP BY operation.

------------------------------------------------------------------------------------------------------------------------------

Q3. Which operators are commonly used with WHERE?

Answer

Comparison Operators (=, >, <, >=, <=, <>)
Logical Operators (AND, OR, NOT)
BETWEEN
IN
LIKE

------------------------------------------------------------------------------------------------------------------------------

Q4. What is the difference between IN and OR?

Answer

Both return the same result, but IN is shorter, cleaner, and easier to maintain when checking multiple values.

------------------------------------------------------------------------------------------------------------------------------


Q5. Is BETWEEN inclusive?

Answer

Yes.

It includes both the starting and ending values.

------------------------------------------------------------------------------------------------------------------------------


Q6. What is the difference between % and _ in the LIKE operator?

Answer

% represents zero or more characters.
_ represents exactly one character.

------------------------------------------------------------------------------------------------------------------------------


Q7. Why should parentheses be used with AND and OR?

Answer

Parentheses improve readability and ensure SQL evaluates conditions in the intended order.
